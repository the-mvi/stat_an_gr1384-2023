## Основные теоретические положения
Случайный вектор $\overrightarrow{\xi} = \left(\xi_1, \dots, \xi_n\right)^\intercal:\Omega\to \mathbb{R}^n$ имеет многомерное нормальное распределение, если существует вектор независимых стандартных нормальных случайных величин $\overrightarrow{\eta} = \left(\eta_1, \dots, \eta_m\right)^\intercal$, матрица A размерности $n \times m$ и вещественный вектор $\overrightarrow{b} = (b_1,\dots,b_n)^\intercal$, такие что:

 $\overrightarrow{\xi} = A\overrightarrow{\eta} + \overrightarrow{b} $.

Плотность многомерного невырожденного нормального распределения имеет вид: $p_{\overrightarrow{\xi}}\left(x_1, \dots, x_n\right) = \frac{1}{(2\pi)^{n/2}\sqrt{\det \Sigma}} \exp\left(-\frac{1}{2}\left(\overrightarrow{x}-\overrightarrow{\mu}\right)^\intercal \Sigma^{-1}\left(\overrightarrow{x}-\overrightarrow{\mu}\right)\right)$

где $\overrightarrow{\mu} = \mathbb{E}\overrightarrow{\xi} = \left(\mathbb{E}\xi_1, \dots, \mathbb{E}\xi_n\right)^\intercal$, $\Sigma = \left\{\sigma_{i,j}\right\}_{i,j=1}^n$, где $\sigma_{i,j} = \mathop{\mathrm{cov}}\left(\xi_i,\xi_j\right)$.

Свойства многомерного нормального распределения:
1. Линейная комбинация случайных переменных, принадлежащих многомерному нормальному распределению, также является нормально распределенной.
2. Условные и безусловные распределения, полученные из многомерного нормального распределения, также являются нормально распределенными.
3. Параметры многомерного нормального распределения полностью описывают его форму.
4. Многомерное нормальное распределение устойчиво относительно линейных преобразований. Если $\overrightarrow{\xi} \sim \mathcal{N}\left(\overrightarrow{\mu}, \Sigma\right)$, $A$ - произвольная матрица размерности $m \times n$, $\overrightarrow{b}$ - произвольный вектор размерности $m \times 1$, то $A\overrightarrow{\xi}+\overrightarrow{b} \sim \mathcal{N}\left(A\overrightarrow{\mu}+\overrightarrow{b}, A \Sigma A^\intercal\right)$.

### Задание 1
Вычислить вектор математических ожиданий и ковариационнные характеристики данного случайного вектора. Построить график плотности распределения $p_{\xi, \eta}$

Преобразуем квадратичную форму в показателе экспоненты

Пусть $q(x,y) = 3x^2 + 4xy + 6y^2 +10x +16y +13 $

$q(x,y) = (3x^2 + 4xy + 10x) + 6y^2 + 16y + 13 = \left(3x^2 + 2x(2y+5)\right) + 6y^2 +16y+13 = 3\left(x^2 + 2x(\frac{2}{3}y + \frac{5}{3}) + (\frac{2}{3}y + \frac{5}{3})^2 - (\frac{2}{3}y + \frac{5}{3})^2\right) + 6y^2 +16y+13 = 3(x + \frac{2}{3}y + \frac{5}{3})^2 - \frac{(2y+5)^2}{3} + 6y^2 + 16y + 13 =  3(x + \frac{2}{3}y + \frac{5}{3})^2 - \frac{4}{3}y^2 - \frac{20}{3}y - \frac{25}{3} + 6y^2 + 16y + 13 = 3(x + \frac{2}{3}y + \frac{5}{3})^2 + \frac{14}{3}y^2 + \frac{28}{3}y + \frac{14}{3} = 3(x + \frac{2}{3}y + \frac{5}{3})^2 + \frac{14}{3}(y+1)^2$

Последнее выражение понадобится для нахождения аффинного преобразования.

$q(x,y)= 3(x + \frac{2}{3}y + \frac{5}{3})^2 + \frac{14}{3}(y+1)^2 = 3\left(x + 1 + \frac{2}{3}(y+1)\right)^2 +\frac{14}{3}(y+1)^2 = 3(x+1)^2 + 4(x+1)(y+1) + 6(y+1)^2$

Найдём вектор мат. ожиданий и обратную ковариационную матрицу.

$\mathbb{E} \begin{pmatrix} \xi \\ \eta \end{pmatrix} = \begin{pmatrix} -1 \\ -1 \end{pmatrix}$

$\Sigma^{-1} = \begin{pmatrix} 3 & 2 \\ 2 & 6 \end{pmatrix}$

Матрица ковариаций

$\Sigma = \frac{1}{14} \begin{pmatrix} 6 & -2 \\ -2 & 3 \end{pmatrix}$

Вычислим дисперсии для каждой случайной величины

$\mathbb{D}\xi = \frac{6}{14} = \frac{3}{7} \;\; \mathbb{D}\eta = \frac{3}{14}$

Найдём ковариацию и коэффициент корреляции

$cov(\xi, \eta) = -\frac{2}{14} = -\frac{1}{7}$

$\rho(\xi, \eta) = -\frac{\frac{1}{7}}{\sqrt{\frac{6}{14} \cdot \frac{3}{14}}} = -\frac{\sqrt{2}}{3}$

Далее найдём C

$C = \frac{1}{2\pi \sqrt{\frac{1}{14}}} = \frac{\sqrt{14}}{2\pi}$

Функция плотности распредления имеет вид:


$p_{\xi, \eta}(x,y) = \frac{\sqrt{14}}{2\pi} \cdot \exp \left(-\frac{1}{2} \begin{bmatrix}\begin{pmatrix} x \\ y \end{pmatrix}  - \begin{pmatrix} -1 \\ -1 \end{pmatrix} \end{bmatrix}^{\intercal} \cdot \begin{pmatrix} 3 & 2 \\ 2 & 6 \end{pmatrix} \cdot \begin{bmatrix}\begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} -1 \\ -1 \end{pmatrix} \end{bmatrix} \right)$

Построим график плотности распределения $p_{\xi, \eta}$


```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

def p(x, y):
    return (np.sqrt(14) / (2 * np.pi)) * np.exp(-0.5 * (3 * x ** 2 + 4 * x * y + 6 * y ** 2 + 10 * x + 16 * y + 13))


x = np.linspace(-3, 1, 200)
y = np.linspace(-3, 2, 200)

X, Y = np.meshgrid(x, y)
Z = p(X, Y)

fig = plt.figure(figsize=(6,6))
ax = plt.axes(projection='3d')

ax.plot_surface(X, Y, Z, cmap='cool')
ax.set_title('График плотности')
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('y', fontsize=12)
ax.set_zlabel('z', fontsize=12)
plt.show()
```

### Задание 3

$ \overrightarrow{s_{1}} = \frac{1}{\sqrt{5}}\begin{pmatrix} -2 \\ 1 \end{pmatrix}$

Объединив векторы и транспонировав матрицу, получим матрицу ортогонального преобразования.

$ Q^{\intercal} = \frac{1}{\sqrt{5}} \begin{pmatrix} -2 & 1 \\ 1 & 2 \end{pmatrix}$

$ Q = \frac{1}{\sqrt{5}} \begin{pmatrix} -2 & 1 \\ 1 & 2 \end{pmatrix}$


$\begin{pmatrix} \xi_{ort} \\ \eta_{ort} \end{pmatrix} = Q \begin{pmatrix} \xi \\ \eta \end{pmatrix} = \frac{1}{\sqrt{5}} \begin{pmatrix} -2\xi + \eta \\ \xi + 2\eta \end{pmatrix}$

$\mathbb{E} \begin{pmatrix} \xi_{ort} \\ \eta_{ort} \end{pmatrix} = \frac{1}{\sqrt{5}} \begin{pmatrix} -2\cdot(-1) + 1\cdot(-1) \\ 1\cdot(-1) + 2\cdot(-1)\end{pmatrix} = \frac{1}{\sqrt{5}} \begin{pmatrix} 1 \\ -3 \end{pmatrix}$

$\Sigma_{ort} = \frac{1}{14} \cdot \frac{1}{5} \begin{pmatrix} -2 & 1 \\ 1 & 2 \end{pmatrix}  \begin{pmatrix} 6 & -2 \\ -2 & 3 \end{pmatrix} \begin{pmatrix} -2 & 1 \\ 1 & 2 \end{pmatrix} = \frac{1}{70} \begin{pmatrix} -14 & 7 \\ 2 & 4 \end{pmatrix} \begin{pmatrix} -2 & 1 \\ 1 & 2 \end{pmatrix} = \frac{1}{70} \begin{pmatrix} 35 & 0 \\ 0 & 10 \end{pmatrix} =  \begin{pmatrix}  \frac{1}{2} & 0 \\ 0 &  \frac{1}{7} \end{pmatrix} $

$C = \frac{1}{2\pi \sqrt{\frac{1}{14}}} = \frac{\sqrt{14}}{2\pi}$

$p_{\xi_{ort}, \eta_{ort}}(x,y) = \frac{\sqrt{14}}{2\pi} \cdot \exp \left(-\frac{1}{2} \begin{bmatrix} \begin{pmatrix} x \\ y \end{pmatrix}  - \begin{pmatrix} \frac{1}{\sqrt{5}} \\ \frac{-3}{\sqrt{5}} \end{pmatrix} \end{bmatrix}^{\intercal} \cdot \begin{pmatrix} 2 & 0 \\ 0 & 7 \end{pmatrix} \cdot \begin{bmatrix}\begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} \frac{1}{\sqrt{5}} \\ \frac{-3}{\sqrt{5}} \end{pmatrix} \end{bmatrix} \right)$

### Задание 4
Вычислить характеристики совместного распределения случайного вектора $(-3\xi- 3\eta, -4\xi - 3\eta)^\intercal$ и записать его плотность. Построить график плотности полученного распределения.

Пусть
$\begin{pmatrix} \xi_{4} \\ \eta_{4} \end{pmatrix} = \begin{pmatrix} -3\xi-3 \eta \\ -4\xi - 3\eta \end{pmatrix}$

$A = \begin{pmatrix} -3 & -3 \\ -4 & -3 \end{pmatrix}$

$\det A = -3 $

Матрица A невырожденная, значит можем вычислить характеристики распределения.

$\begin{pmatrix} \xi_{4} \\ \eta_{4} \end{pmatrix} = \begin{pmatrix} -3 & -3 \\ -4 & -3 \end{pmatrix} \begin{pmatrix} \xi \\ \eta \end{pmatrix}$

$\mathbb{E} \begin{pmatrix} \xi_{4} \\ \eta_{4} \end{pmatrix} = \begin{pmatrix} -3 & -3 \\ -4 & -3 \end{pmatrix} \begin{pmatrix} -1 \\ -1 \end{pmatrix} = \begin{pmatrix} 6 \\ 7 \end{pmatrix}$

$\Sigma_{4} = \frac{1}{14} \begin{pmatrix}3 & 3 \\ 4 & 3 \end{pmatrix}  \begin{pmatrix} 6 & -2 \\ -2 & 3 \end{pmatrix} \begin{pmatrix} 3 & 4 \\ 3 & 3  \end{pmatrix} = \frac{1}{14}\begin{pmatrix} 12 & 3 \\ 18 & 1 \end{pmatrix} \begin{pmatrix} 3 & 4 \\ 3 & 3  \end{pmatrix} = \frac{1}{14}\begin{pmatrix} 45 & 57 \\ 57 & 75 \end{pmatrix} = \begin{pmatrix} \frac{45}{14} & \frac{57}{14} \\ \frac{57}{14} & \frac{75}{14} \end{pmatrix}$

$\Sigma_{4}^{-1} = \begin{pmatrix} \frac{25}{3} & -\frac{19}{3} \\ -\frac{19}{3} & 5 \end{pmatrix}$

$\mathbb{D}\xi_{4} = \frac{45}{14} \;\; \mathbb{D}\eta_{4} = \frac{75}{14}$

Найдём ковариацию и коэффициент корреляции

$cov(\xi, \eta) = \frac{57}{14}$

$\rho(\xi, \eta) = \frac{\frac{57}{14}}{\sqrt{\frac{45}{14} \cdot \frac{75}{14}}} = \frac{19}{5\sqrt{15}}$

Далее найдём C

$C = \frac{1}{2\pi \sqrt{\frac{45\cdot 75 - 57^2}{14^2}}} = \frac{1}{2\pi \sqrt{\frac{126}{196}}} = \frac{\sqrt{14}}{6\pi}$

Функция плотности распредления имеет вид:


$p_{\xi_{4}, \eta_{4}}(x,y) = \frac{\sqrt{14}}{6\pi} \cdot \exp \left(-\frac{1}{2} \begin{bmatrix}\begin{pmatrix} x \\ y \end{pmatrix}  - \begin{pmatrix} 6 \\ 7 \end{pmatrix} \end{bmatrix}^{\intercal} \cdot \begin{pmatrix} \frac{25}{3} & -\frac{19}{3} \\ -\frac{19}{3} & 5 \end{pmatrix} \cdot \begin{bmatrix}\begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} 6 \\ 7 \end{pmatrix} \end{bmatrix} \right)$

Построим график плотности распределения $p_{\xi_{4}, \eta_{4}}$


```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

def p(x, y):
    return ( np.sqrt(14) / (6 * np.pi)) * np.exp(-0.5 * (1 / 3) * (25*x**2 - 34*x - 38*x*y + 15*y**2 + 18*y + 39))


x = np.linspace(0, 12, 200)
y = np.linspace(-10, 20, 200)

X, Y = np.meshgrid(x, y)
Z = p(X, Y)

fig = plt.figure(figsize=(10,8))
ax = plt.axes(projection='3d')

ax.plot_surface(X, Y, Z, cmap='cool')
ax.set_title('График плотности')
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('y', fontsize=12)
ax.set_zlabel('z', fontsize=12)
plt.show()
```
