## Основные теоретические положения
Случайный вектор $\overrightarrow{\xi} = \left(\xi_1, \dots, \xi_n\right)^\intercal:\Omega\to \mathbb{R}^n$ имеет многомерное нормальное распределение, если существует вектор независимых стандартных нормальных случайных величин $\overrightarrow{\eta} = \left(\eta_1, \dots, \eta_m\right)^\intercal$, матрица A размерности $n \times m$ и вещественный вектор $\overrightarrow{b} = (b_1,\dots,b_n)^\intercal$, такие что: 

 $\overrightarrow{\xi} = A\overrightarrow{\eta} + \overrightarrow{b} $.

Плотность многомерного невырожденного нормального распределения имеет вид: $p_{\overrightarrow{\xi}}\left(x_1, \dots, x_n\right) = \frac{1}{(2\pi)^{n/2}\sqrt{\det \Sigma}} \exp\left(-\frac{1}{2}\left(\overrightarrow{x}-\overrightarrow{\mu}\right)^\intercal \Sigma^{-1}\left(\overrightarrow{x}-\overrightarrow{\mu}\right)\right)$

где $\overrightarrow{\mu} = \mathbb{E}\overrightarrow{\xi} = \left(\mathbb{E}\xi_1, \dots, \mathbb{E}\xi_n\right)^\intercal$, $\Sigma = \left\{\sigma_{i,j}\right\}_{i,j=1}^n$, где $\sigma_{i,j} = \mathop{\mathrm{cov}}\left(\xi_i,\xi_j\right)$.

Свойства многомерного нормального распределения:
1. Линейная комбинация случайных переменных, принадлежащих многомерному нормальному распределению, также является нормально распределенной.
2. Условные и безусловные распределения, полученные из многомерного нормального распределения, также являются нормально распределенными.
3. Параметры многомерного нормального распределения полностью описывают его форму.
4. Многомерное нормальное распределение устойчиво относительно линейных преобразований. Если $\overrightarrow{\xi} \sim \mathcal{N}\left(\overrightarrow{\mu}, \Sigma\right)$, $A$ - произвольная матрица размерности $m \times n$, $\overrightarrow{b}$ - произвольный вектор размерности $m \times 1$, то $A\overrightarrow{\xi}+\overrightarrow{b} \sim \mathcal{N}\left(A\overrightarrow{\mu}+\overrightarrow{b}, A \Sigma A^\intercal\right)$.

### Задание 1
Вычислить вектор математических ожиданий и ковариационнные характеристики данного случайного вектора. Построить график плотности распределения $p_{\xi, \eta}$

Воспользуемся методом Лагранжа приведения квадратичной формы к каноническому виду.

Пусть $q(x,y) = 6x^2 -4xy +9y^2 +8x +14y +11 $

$q(x,y) = \left(6x^2-4xy+8x\right) + 9y^2 + 14y + 11  = \left(6x^2 - 2x(2y-4)\right) + 9y^2 + 14y + 11
= 6 \left( x^2 - 2x\cdot \frac{1}{3}(y-2) \right) + 9y^2 + 14y + 11 = 6\left(x^2 - 2x \left(\frac{y}{3} -\frac{2}{3}\right) + \left(\frac{y}{3} -\frac{2}{3} \right)^2 - \left(\frac{y}{3} -\frac{2}{3} \right)^2 \right) + 9y^2 + 14y + 11 = 6\left( x - \frac{y}{3} + \frac{2}{3} \right)^2 - 6\left( \frac{y}{3} -\frac{2}{3} \right)^2 + 9y^2 + 14y + 11= 6 \left( x - \frac{y}{3} + \frac{2}{3} \right)^2 + \frac{25}{3}(y+1)^2 $

Запомним последнее выражение для нахождения аффинного преобразования во втором задании.

$q(x,y)= 6 \left( -\frac{1}{3}(y+1) + x + 1 \right)^2 + \frac{25}{3}(y+1)^2 = 6 \left( (x+1)^2 -\frac{2}{3}(y+1)(x+1) + \frac{1}{9}(y+1)^2 \right) + \frac{25}{3}(y+1)^2 = 6(x+1)^2 -4(y+1)(x+1) + 9(y+1)^2$

Тогда найдём вектор мат. ожиданий и обратную ковариационную матрицу.

$\mathbb{E} \begin{pmatrix} \xi \\ \eta \end{pmatrix} = \begin{pmatrix} -1 \\ -1 \end{pmatrix}$

$\Sigma^{-1} = \begin{pmatrix} 6 & -2 \\ -2 & 9 \end{pmatrix}$

Матрица ковариаций 

$\Sigma = \frac{1}{50} \begin{pmatrix} 9 & 2 \\ 2 & 6 \end{pmatrix}$

Вычислим дисперсии для каждой случайной величины

$\mathbb{D}\xi = \frac{9}{50} \;\; \mathbb{D}\eta = \frac{6}{50} = \frac{3}{25}$

Найдём ковариацию и коэффициент корреляции

$cov(\xi, \eta) = \frac{1}{25}$

$\rho(\xi, \eta) = \frac{\frac{1}{25}}{\sqrt{\frac{6}{50} \cdot \frac{9}{50}}} = \frac{\sqrt{6}}{9}$

Далее найдём C

$C = \frac{1}{2\pi \sqrt{\frac{1}{50}}} = \frac{5}{\pi\sqrt{2}}$

Функция плотности распредления имеет вид:


$p_{\xi, \eta}(x,y) = \frac{5}{\pi\sqrt{2}} \cdot \exp \left(-\frac{1}{2} \begin{bmatrix}\begin{pmatrix} x \\ y \end{pmatrix}  - \begin{pmatrix} -1 \\ -1 \end{pmatrix} \end{bmatrix}^{\intercal} \cdot \begin{pmatrix} 6 & -2 \\ -2 & 9 \end{pmatrix} \cdot \begin{bmatrix}\begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} -1 \\ -1 \end{pmatrix} \end{bmatrix} \right)$

Построим график плотности распределения $p_{\xi, \eta}$


```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

def p(x, y):
    return (5 / (np.pi * np.sqrt(2))) * np.exp(-0.5 * (6 * x ** 2 - 4 * x * y + 9 * y ** 2 + 8 * x + 14 * y + 11))


x = np.linspace(-3, 1, 200)
y = np.linspace(-3, 2, 200)

X, Y = np.meshgrid(x, y)
Z = p(X, Y)

fig = plt.figure(figsize=(10,8))
ax = plt.axes(projection='3d')

ax.plot_surface(X, Y, Z, cmap='cool')
ax.set_title('График плотности')
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('y', fontsize=12)
ax.set_zlabel('z', fontsize=12)
plt.show()
```

### Задание 3
Найти ортогональное преобразование, переводящее соответствующий центрированный случайный вектор в вектор с независимыми компонентами. Построить график плотности полученного распределения.

$\Sigma^{-1} = \begin{pmatrix} 6 & -2 \\ -2 & 9 \end{pmatrix}$

Для этого найдём собственные числа

Получим нормированный собственный вектор:

$ \overrightarrow{s_{1}} = \frac{1}{\sqrt{5}}\begin{pmatrix} 2 \\ 1 \end{pmatrix}$

$ \lambda_{2} = 10$

Получим нормированный собственный вектор:

$ \overrightarrow{s_{2}} = \frac{1}{\sqrt{5}}\begin{pmatrix} -1 \\ 2 \end{pmatrix}$

Объединив векторы и транспонировав матрицу, получим матрицу ортогонального преобразования.

$ Q^{\intercal} = \frac{1}{\sqrt{5}} \begin{pmatrix} 2 & -1 \\ 1 & 2 \end{pmatrix}$

$ Q = \frac{1}{\sqrt{5}} \begin{pmatrix} 2 & 1 \\ -1 & 2 \end{pmatrix}$


$\begin{pmatrix} \xi_{ort} \\ \eta_{ort} \end{pmatrix} = Q \begin{pmatrix} \xi \\ \eta \end{pmatrix} = \frac{1}{\sqrt{5}} \begin{pmatrix} 2\xi + \eta \\ -\xi + 2\eta \end{pmatrix}$

$\mathbb{E} \begin{pmatrix} \xi_{ort} \\ \eta_{ort} \end{pmatrix} = \frac{1}{\sqrt{5}} \begin{pmatrix} 2\cdot(-1) + (-1) \\ 1 + 2\cdot(-1) \end{pmatrix} = \frac{1}{\sqrt{5}} \begin{pmatrix} -3 \\ -1 \end{pmatrix}$

$\Sigma_{ort} = \frac{1}{50} \cdot \frac{1}{5} \begin{pmatrix} 2 & 1 \\ -1 & 2 \end{pmatrix}  \begin{pmatrix} 9 & 2 \\ 2 & 6 \end{pmatrix} \begin{pmatrix} 2 & -1 \\ 1 & 2 \end{pmatrix} = \frac{1}{250} \begin{pmatrix} 20 & 10 \\ -5 & 10 \end{pmatrix} \begin{pmatrix} 2 & -1 \\ 1 & 2 \end{pmatrix} = \frac{1}{250} \begin{pmatrix} 50 & 0 \\ 0 & 25 \end{pmatrix} =  \begin{pmatrix}  \frac{1}{5} & 0 \\ 0 &  \frac{1}{10} \end{pmatrix} $

$\Sigma^{-1}_{ort} = \begin{pmatrix} 5 & 0 \\ 0 & 10 \end{pmatrix}$

$\mathbb{D}\xi_{ort} = \frac{1}{5} \;\; \mathbb{D}\eta_{ort} = \frac{1}{10} $

$cov(\xi_{ort}, \eta_{ort}) = 0 \;\; \rho(\xi_{ort}, \eta_{ort}) = 0$

Далее найдём C

$C = \frac{1}{2\pi \sqrt{\frac{1}{50}}} = \frac{5}{\pi\sqrt{2}}$


Функция плотности распредления имеет вид:


$p_{\xi_{ort}, \eta_{ort}}(x,y) = \frac{5}{\pi\sqrt{2}} \cdot \exp \left(-\frac{1}{2} \begin{bmatrix} \begin{pmatrix} x \\ y \end{pmatrix}  - \begin{pmatrix} \frac{-3}{\sqrt{5}} \\ \frac{-1}{\sqrt{5}} \end{pmatrix} \end{bmatrix}^{\intercal} \cdot \begin{pmatrix} 5 & 0 \\ 0 & 10 \end{pmatrix} \cdot \begin{bmatrix}\begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} \frac{-3}{\sqrt{5}} \\ \frac{-1}{\sqrt{5}} \end{pmatrix} \end{bmatrix} \right)$

### Задание 4
Вычислить характеристики совместного распределения случайного вектора $(4\xi- \eta, \xi - 5\eta)^\intercal$ и записать его плотность. Построить график плотности полученного распределения.

Пусть
$\begin{pmatrix} \xi_{4} \\ \eta_{4} \end{pmatrix} = \begin{pmatrix} 4\xi- \eta \\ \xi - 5\eta \end{pmatrix}$

$A = \begin{pmatrix} 4 & -1 \\ 1 & -5 \end{pmatrix}$

$\det A = -19 $ 

Матрица A невырожденная, значит можем вычислить характеристики распределения.

$\begin{pmatrix} \xi_{4} \\ \eta_{4} \end{pmatrix} = \begin{pmatrix} 4 & -1 \\ 1 & -5 \end{pmatrix} \begin{pmatrix} \xi \\ \eta \end{pmatrix}$

$\mathbb{E} \begin{pmatrix} \xi_{4} \\ \eta_{4} \end{pmatrix} = \begin{pmatrix} 4 & -1 \\ 1 & -5 \end{pmatrix} \begin{pmatrix} -1 \\ -1 \end{pmatrix} = \begin{pmatrix} -3 \\ 4 \end{pmatrix}$ 

$\Sigma_{4} = \frac{1}{50} \begin{pmatrix}4 & -1 \\ 1 & -5 \end{pmatrix}  \begin{pmatrix} 9 & 2 \\ 2 & 6 \end{pmatrix} \begin{pmatrix} 4 & 1 \\ -1 & -5  \end{pmatrix} = \frac{1}{50}\begin{pmatrix} 34 & 2 \\ -1 & -28 \end{pmatrix} \begin{pmatrix} 4 & 1 \\ -1 & -5  \end{pmatrix} = \frac{1}{50}\begin{pmatrix} 134 & 24 \\ 24 & 139 \end{pmatrix} = \begin{pmatrix} \frac{67}{25} & \frac{12}{25} \\ \frac{12}{25} & \frac{139}{50} \end{pmatrix}$

$\Sigma_{4}^{-1} = \frac{1}{361} \begin{pmatrix} 139 & -24 \\ -24 & 134 \end{pmatrix} = \begin{pmatrix} \frac{139}{361} & -\frac{24}{361} \\ -\frac{24}{361} & \frac{134}{361} \end{pmatrix}$

$\mathbb{D}\xi_{4} = \frac{67}{25} \;\; \mathbb{D}\eta_{4} = \frac{139}{50}$

Найдём ковариацию и коэффициент корреляции

$cov(\xi, \eta) = \frac{12}{25}$

$\rho(\xi, \eta) = \frac{\frac{12}{25}}{\sqrt{\frac{67}{25} \cdot \frac{139}{50}}} = \frac{12\sqrt{2}}{\sqrt{9313}}$

Далее найдём C

$C = \frac{1}{2\pi \sqrt{\frac{134\cdot 139 - 24^2}{50^2}}} = \frac{1}{2\pi \sqrt{\frac{361}{50}}} = \frac{5\sqrt{2}}{38\pi}$

Функция плотности распредления имеет вид:


$p_{\xi_{4}, \eta_{4}}(x,y) = \frac{5\sqrt{2}}{38\pi} \cdot \exp \left(-\frac{1}{2} \begin{bmatrix}\begin{pmatrix} x \\ y \end{pmatrix}  - \begin{pmatrix} -3 \\ 4 \end{pmatrix} \end{bmatrix}^{\intercal} \cdot \begin{pmatrix} \frac{139}{361} & -\frac{24}{361} \\ -\frac{24}{361} & \frac{134}{361} \end{pmatrix} \cdot \begin{bmatrix}\begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} -3 \\ 4 \end{pmatrix} \end{bmatrix} \right)$

Построим график плотности распределения $p_{\xi_{4}, \eta_{4}}$


```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

def p(x, y):
    return (5* np.sqrt(2) / (38 * np.pi)) * np.exp(-0.5 * (1/361) * (139*x**2 +1026*x + 3971 - 48*x*y -1216*y +134*y**2))


x = np.linspace(-9, 7, 200)
y = np.linspace(-15, 20, 200)

X, Y = np.meshgrid(x, y)
Z = p(X, Y)

fig = plt.figure(figsize=(10,8))
ax = plt.axes(projection='3d')

ax.plot_surface(X, Y, Z, cmap='cool')
ax.set_title('График плотности')
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('y', fontsize=12)
ax.set_zlabel('z', fontsize=12)
plt.show()
```
