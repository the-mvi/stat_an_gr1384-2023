## Основные теоретические положения
Случайный вектор $\overrightarrow{\xi} = \left(\xi_1, \dots, \xi_n\right)^\intercal:\Omega\to \mathbb{R}^n$ имеет многомерное нормальное распределение, если $\exists$ вектор независимых стандартных нормальных случайных величин $\overrightarrow{\eta} = \left(\eta_1, \dots, \eta_m\right)^\intercal$, матрица A размерности $n \times m$ и вещественный вектор $\overrightarrow{b} = (b_1,\dots,b_n)^\intercal$: 
 $$
 \overrightarrow{\xi} = A\overrightarrow{\eta} + \overrightarrow{b} 
 $$

Многомерное случайное распределение - абсолютно непрерывное распределение, с плотностью:
$$
\mathcal{p}_{{\xi}}(x) = \frac{1}{(2\pi)^\frac{n}{2}\sqrt{\det \Sigma}} \exp\left(-\frac{1}{2}\begin{bmatrix} {x}-{\mu}\end{bmatrix} ^\intercal \Sigma^{-1}\begin{bmatrix}{x}-{\mu}\end{bmatrix}\right)
$$
Его числовые характеристики:
Математическое ожидание: $\mathbb{E}\overrightarrow{\xi}=\mu$
Матрица ковариаций: $\left\{\sigma_{i,j}\right\}_{i,j=1}^n = \Sigma$, где $\sigma_{i,j} = \mathop{\mathrm{cov}}\left(\xi_i,\xi_j\right)$


Свойства многомерного нормального распределения:
1. Линейная комбинация случайных переменных также является нормально распределенной.
2. Условные и безусловные распределения также являются нормально распределенными.
3. Параметры многомерного нормального распределения полностью описывают его форму.
4. Устойчивость относительно линейных преобразований. Если $\overrightarrow{\xi} \sim \mathcal{N}\left(\overrightarrow{\mu}, \Sigma\right)$, $A$ - произвольная матрица размерности $m \times n$, $\overrightarrow{b}$ - произвольный вектор размерности $m \times 1$, то $A\overrightarrow{\xi}+\overrightarrow{b} \sim \mathcal{N}\left(A\overrightarrow{\mu}+\overrightarrow{b}, A \Sigma A^\intercal\right)$.

### Задание 1
Вычислить вектор математических ожиданий и ковариационнные характеристики данного случайного вектора. Построить график плотности распределения $p_{\xi, \eta}$

Введем функцию, равную квадратичной форме в показателе экспоненты:
$$
q(x, y) = 7x^2 + 8xy + 13y^2 + 6x - 18y + 12
$$

Преобразуем ее по методу Лагранжа:
$$
q(x, y) = (7x^2 + 2x(4y + 3)) + 13y^2 - 18y + 12 = \\
    7 \left(x^2 + 2x \left(\frac{4}{7}y + \frac{3}{7}\right) \right) + 13y^2 - 18y + 12 = \\
    7 \left(x^2 + 2x\left(\frac{4}{7}y + \frac{3}{7}\right) + \left(\frac{4}{7}y + \frac{3}{7}\right)^2\right) - \frac{16}{7}y^2 - \frac{24}{7}y - \frac{9}{7} + 13y^2 - 18y + 12 = \\
    7 \left(x + \frac{4}{7}y + \frac{3}{7}\right)^2 + \frac{1}{7}(75y^2 - 150y + 75) = \\
    7 \left(x + \frac{4}{7}y + \frac{3}{7}\right)^2 + \frac{75}{7}(y-1)^2
$$
Полученное выражение будет полезно в дальнейших вычислениях (см. задание 2)

Продолжим преобразование:
$$
q(x, y) = 7\left(x + \frac{4}{7}y + \frac{3}{7}\right)^2 + \frac{75}{7}(y-1)^2 = \\
    7\left(x + 1 + \frac{4}{7}(y - 1)\right)^2 + \frac{75}{7}(y-1)^2 = \\
    7(x + 1)^2 + 8(x+1)\cdot(y-1) + 13(y-1)^2
$$

Полученные коэффициенты образуют матрицу, обратную матрице ковариаций:
$$
\Sigma^{-1} = \begin{pmatrix}
  7 & 4 \\
  4 & 13
\end{pmatrix}
$$
Ее определитель будет равен $\det\left(\Sigma^{-1}\right) = 75$

Таким образом, получим матрицу ковариаций:
$$
\Sigma = \frac{1}{75} \cdot \begin{pmatrix}
  13 & -4 \\
  -4 & 7
\end{pmatrix}
$$

Помимо этого, полученное по методу Лагранжа выражение позволяет найти вектор математического ожидания:
$$
\mu = \mathbb{E}_{\xi,\eta} = \begin{pmatrix} \xi\\\eta \end{pmatrix} = \begin{pmatrix} -1\\1 \end{pmatrix}
$$

Из матрицы ковариаций выделим дисперсии каждой случайной величины:
$$
\mathbb{D}\xi = \frac{13}{75} \\ \mathbb{D}\eta = \frac{7}{75}
$$
Аналогично выделим ковариацию между случайными величинами:
$$
cov\left(\xi, \eta\right) = -\frac{4}{75}
$$
Откуда найдем коэффициент корреляции:
$$
\rho(\xi, \eta) = \frac{-\frac{4}{75}}{\sqrt{\frac{13}{75} \cdot \frac{7}{75}}} = -\frac{4}{\sqrt{91}}
$$

Также найдем константу $C$:
$$
C = \frac{1}{(2\pi)^{\frac{n}{2}} \sqrt{det\Sigma}} = \frac{1}{2\pi\sqrt{\frac{1}{75}}} = \frac{5\sqrt{3}}{2\pi}
$$

В итоге получили следующую формулу плотности распределения:

$$
p_{\xi, \eta}(x,y) = \frac{5\sqrt{3}}{2\pi} \cdot \exp \left(-\frac{1}{2} \begin{bmatrix}\begin{pmatrix} x \\ y \end{pmatrix}  - \begin{pmatrix} -1 \\ 1 \end{pmatrix} \end{bmatrix}^{\intercal} \cdot \begin{pmatrix} 7 & 4 \\ 4 & 13 \end{pmatrix} \cdot \begin{bmatrix}\begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} -1 \\ 1 \end{pmatrix} \end{bmatrix} \right)
$$


```python
import numpy as np
import matplotlib.pyplot as plt


def p(x, y, sigma, mean):
    vec = np.array([x, y], dtype=float)
    inv = np.linalg.inv(sigma)
    det = np.linalg.det(sigma)
    C = 1 / (2 * np.pi * np.sqrt(det))
    return C * np.exp(-1 / 2 * (vec - mean).T.dot(inv).dot(vec - mean))


def draw(sigma, mean, bound_x=(-2.5, 0.5), bound_y=(0, 2), count=200):
    x_space = np.linspace(bound_x[0], bound_x[1], count)
    y_space = np.linspace(bound_y[0], bound_y[1], count)

    X, Y = np.meshgrid(x_space, y_space)
    Z = np.zeros(X.shape)
    for i in range(X.shape[0]):
        for j in range(X.shape[1]):
            Z[i, j] = p(X[i, j], Y[i, j], sigma, mean)

    fig = plt.figure()
    ax = fig.add_subplot(projection='3d')
    ax.plot_surface(X, Y, Z, cmap='viridis')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    plt.title('График плотности')

    plt.show()

```

```python
draw(1/75*np.array([[13, -4], [-4, 7]], dtype=float), np.array([-1, 1], dtype=float))
```

### Задание 3
Найти ортогональное преобразование, переводящее соответствующий центрированный случайный вектор в вектор с независимыми компонентами. Построить график плотности полученного распределения.

Найдем собственные числа матрицы $\Sigma^{-1}$
$$
\begin{vmatrix} 7 - \lambda & 4 \\ 4 & 13 - \lambda \end{vmatrix} = 0 \Leftrightarrow \\
\lambda^2 - 20\lambda + 75 = 0 \Leftrightarrow \\
\lambda_1 = 5 \\
\lambda_2 = 15
$$

Отсюда найдем собственные вектора:
$$
\begin{pmatrix} 2 & 4 \\ 4 & 8 \end{pmatrix} \cdot v_1 = 0 \Rightarrow \\
v_1 = \frac{1}{\sqrt5}\begin{pmatrix}
  2 \\
  -1
\end{pmatrix} 
$$
$$
\begin{pmatrix} -8 & 4 \\ 4 & 2 \end{pmatrix} \cdot v_1 = 0 \Rightarrow \\
v_2 = \frac{1}{\sqrt5}\begin{pmatrix}
  1 \\
  2
\end{pmatrix}
$$

Отсюда получим матрицу преобразования, транспонировав объединение векторов:
$$
Q^{\intercal} = \frac{1}{\sqrt5} \begin{pmatrix}
  2 & 1 \\
  -1 & 2
\end{pmatrix} \Rightarrow \\
Q = \frac{1}{\sqrt5} \begin{pmatrix}
  2 & -1 \\
  1 & 2
\end{pmatrix}
\\ \\
\begin{pmatrix}
  \xi_{ort} \\
  \eta_{ort}
\end{pmatrix} = 
Q \begin{pmatrix}
  \xi \\
  \eta
\end{pmatrix} =
\frac{1}{\sqrt5}
\begin{pmatrix}
  2\xi - \eta \\
  \eta + 2\eta
\end{pmatrix}
$$


Проверим, что компоненты вектора независимы:
$$
\Sigma_{ort} = Q \Sigma Q^{\intercal} =
\frac{1}{5 \cdot 75}
\begin{pmatrix}
  2 & -1 \\
  1 & 2
\end{pmatrix} 
\begin{pmatrix}
  13 & -4 \\
  -4 & 7
\end{pmatrix} 
\begin{pmatrix}
  2 & 1 \\
  -1 & 2
\end{pmatrix} 
=
\begin{pmatrix}
  \frac{1}{5} & 0 \\
  0 & \frac{1}{15}
\end{pmatrix}
\\ \\
\Sigma_{ort}^{-1} = 
\begin{pmatrix}
  \lambda_1 & 0 \\
  0 & \lambda_2
\end{pmatrix}
$$

Для построения графика необходимо математическое ожидание:

$$
\mathbb E
\begin{pmatrix}
  \xi_{ort} \\
  \eta_{ort}
\end{pmatrix} =
\frac{1}{\sqrt 5}
\begin{pmatrix}
  2\mathbb E \xi - \mathbb E\eta \\
  \mathbb E\xi + 2\mathbb E \eta

### Задание 5
Найти условное распределение $\xi$ при условии $\eta$.
Вычислить E($\xi|\eta$) и D($\xi|\eta$).

Следуя из свойства ниже, каждая компонента будет нормально распределена
> Условные и безусловные распределения, полученные из многомерного нормального распределения, также являются нормально распределенными.

Из пункта 1: $\mathbb{E}\eta = 1$ $\mathbb{D}\eta = \frac{7}{75}$
$$
p_{\eta}(y) = \frac{1}{\sqrt{2\pi \frac{7}{75}}}\cdot \exp \left(\frac{-(y - 1)^2}{2 \frac{7}{75}} \right) =

$$

По формуле $p_{\xi|\eta=y}(x) = \frac{p_{\xi,\eta}(x,y)}{p_{\eta}(y)}$

$$
p_{\xi|\eta=y}(x) = \frac{\frac{5\sqrt{3}}{2\pi} \cdot \exp \left( -\frac{1}{2}(7x^2 + 8xy + 13y^2 + 6x - 18y + 12) \right)}{\frac{5\sqrt3}{\sqrt{14\pi}}\cdot \exp \left(\frac{-(y - 1)^2}{2 \cdot \frac{7}{75}} \right)} = \\ 
\frac{1}{\sqrt{2\pi} \cdot \frac{1}{\sqrt{7}}} \cdot \exp \left(-\frac{7}{2} \cdot \left(x^2 + \frac{8}{7}xy + \frac{6}{7}x + \frac{13}{7}y^2 + \frac{-75y^2+24y+9}{49}\right)\right)
$$
 
Отсюда $\mathbb{D}(\xi|\eta=y) = \frac{1}{7}$
Рассмотрим квадратичную форму
$$
x^2 + \frac{8}{7}xy + \frac{6}{7}x + \frac{13}{7}y^2 + \frac{-75y^2+24y+9}{49} = \\ 
x^2 + 2x \left(\frac{4}{7}y + \frac{3}{7}\right) + \left(\frac{4}{7}y + \frac{3}{7}\right)^2 + \frac{75}{49}y^2 - \frac{24}{49}y - \frac{9}{49} + \frac{-75y^2+24y+9}{49} = \\
\left(x - \left(- \frac{4}{7}y - \frac{3}{7}\right)\right)^2
$$

Значит, $\mathbb{E}(\xi|\eta=y) = - \frac{4}{7}y - \frac{3}{7}$
