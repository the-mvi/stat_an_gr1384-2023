## Основные теоретические положения

- **Случайная величина**. Пусть $(\Omega, F, P)$ - вероятностный эксперимент. $\xi$ - случайная величина, если $\xi$ - вещественная измеримая функция (или отображение), определенной на множестве элементарных событий, т.е. $\xi:\Omega\rightarrow\mathbb{R}$, и функция $\xi$ является измеримой.
- **Случайный вектор**. Пусть $\xi_{1},\xi_{2},...,\xi_{d}$ - случайные величины, определенные на одном и том же вероятностном пространстве. Определим отображение: $$\xi=(\xi_{1},\xi_{2},...,\xi_{d}):\Omega\rightarrow\mathbb{R}^d$$ по правилу $\xi(\omega)=(\xi_{1}(\omega),\xi_{2}(\omega),...,\xi_{d}(\omega))$. Это и есть d-мерный случайный вектор. 
- **Математическим ожиданием** случайной величины $\xi$ называют **число** $\mathbb{E}\xi$ определяемое равенством $$\mathbb{E}\xi = \int_{\omega\in\Omega} \xi(\omega) \, dP(\omega) = \int_{-\infty}^{\infty} x \, dF_{\xi}(x)$$ где $F\substack(\xi)$ - функция распределения величины $\xi$. Если интеграл расходится, то говорят, что у случайной величины отсутствует математическое ожидание. Для дискретной случайной величины $\mathbb{E}\xi=\sum_{i;p_{i}>0}a_{i}p_{i}, p_{i}=P(\xi = a_{i}), \sum_{i}p_{i} = 1$ Для абсолютно непрерывной случайной величины $\mathbb{E}\xi = \int_{\mathbb{R}} xp_{\xi}(x) \, dx$.
- **Дисперсией** случайной величины $\xi$ называется центральный момент второго порядка $\mathbb{E}(\xi-\mathbb{E}\xi)^2$, если этот момент существует.
<br>Стандартное обозначение для дисперсии - $\mathbb{D}\xi$.
<br>Из определения математического ожидания следует, что $$\mathbb{D}\xi = \int_{\Omega} (\xi(\omega)-\mathbb{E}\xi)^2 \, dP(\omega)$$
Таким образом, дисперсия - это среднее значение квадрата отклонения случайной величины от её математического ожидания.
- **Ковариационная характеристика** используются для описания связи или зависимости между случайными величинами. Они помогают измерить, насколько две случайные величины варьируются вместе. Ниже приведены основные ковариационные характеристики:
    - **Ковариация**. Пусть $(\Omega, F, P)$ - вероятностный эксперимент, а $\xi, \eta$ - случайные величины, определенные в нем. Смешанный центральный момент второго порядка $$\mathbb{E}(\xi-\mathbb{E}\xi)(\eta-\mathbb{E}\eta) = \mathop{\mathrm{cov}}\limits(\xi, \eta)$$ называется ковариацией случайных величин $\xi$ и $\eta$.
    - **Коэффициент корреляции**. Пусть $\xi,\eta$ - случайные величины, у которых существуют дисперсии и $\mathbb{D}\xi\mathbb{D}\eta > 0$. Тогда число $$\rho(\xi,\eta) = \frac{\mathop{\mathrm{cov}}\limits(\xi, \eta)}{\sqrt{\mathbb{D}\xi\cdot\mathbb{D}\eta}}$$
    - **Матрица ковариациий** - это матрица, составленная из попарных ковариаций элементов одного или двух случайных векторов. Элементы на главной диагонали представляют собой дисперсии каждой компоненты вектора.
- **Стандартный нормальный вектор** - это случайный вектор, компоненты которого независимы и имеют стандартное нормальное распределение. В двумерном случае стандартный нормальный вектор имеет стандартное нормальное двумерное распределение (распределение Гаусса или нормальное распределение).
Каждая компонента стандартного нормального вектора имеет следующие характеристики:
1. Математическое ожидание каждой компоненты вектора равно нулю
2. Дисперсия каждой компоненты вектора равно единице.
3. Матрица ковариаций представляет из себя единичную матрицу.

**1.** Вычислить вектор математических ожиданий и ковариационные характеристики данного случайного вектора. Построить график плотности распределения $p\tiny\xi,\eta$.

Преобразуем квадратичную форму в показателе экспоненты

$q(x,y) = 2x^2 + 4xy + 5y^2 - 8x - 14y + 11$

$2x^2 + 4xy + 5y^2 - 8x - 14y + 11$
<br>$= (2x^2 + 4xy - 8x) + 5y^2 - 14y + 11$
<br>$= 2(x^2 + 2xy - 4x) + 5y^2 - 14y + 11$
<br>$= 2(x^2 + 2x(y - 2) + (y-2)^2) - 2y^2 + 8y - 8 + 5y^2 - 14y + 11$
<br>$= 2(x + (y - 2))^2 + 3y^2 - 6y + 3$
<br>$= 2(x + (y - 2))^2 + 3(y - 1)^2$
<br>$= 2((x - 1) + (y - 1))^2 + 3(y - 1)^2$

$= 2(x - 1)^2 + 4(x - 1)(y - 1) + 5(y - 1)^2$

В последней строке коэффициенты перед $(x-1)^2$, $(x-1)(y-1)$, $(y-1)^2$ принимаются в обратную матрицу ковариаций, причём 4 разбивается на две 2.
$$\Sigma^{-1} = \begin{pmatrix}
    2 & 2\\
    2 & 5
    \end{pmatrix}$$ определитель которого равен 6: $\det\Sigma^{-1} = 2\cdot5 - 2\cdot2 = 6$
Далее получаем матрицу ковариаций следующим образом: значения на главной диагонали меняем местами на главной диагонали, а значения на побочной диагонали изменяют знак; также каждый элемент матрицы делится на определитель. 
$$\Sigma = \frac{1}{6}\begin{pmatrix}
    5 & -2\\
    -2 & 2
    \end{pmatrix}$$

Также от полученного выражения получаем вектор математического ожидания $\mu = \mathbb{E}_{\xi,\eta} = \begin{pmatrix} \xi\\\eta \end{pmatrix} = \begin{pmatrix} 1\\1 \end{pmatrix}$
Из матрицы ковариации на главной диагонали располагаются дисперсии $\mathbb{D}_{\xi,\eta} = \begin{pmatrix} \mathbb{D}_{\xi}\\\mathbb{D}_{\eta} \end{pmatrix} = \begin{pmatrix} \frac{5}{6}\\\frac{1}{3} \end{pmatrix}$, на побочной диагонали - ковариация $\mathop{\mathrm{cov}}\limits(\xi, \eta) = -\frac{1}{3}$.
<br>Далее получается коэффициент корреляции $\rho(\xi,\eta) = \frac{\mathop{\mathrm{cov}}\limits(\xi, \eta)}{\sqrt{\mathbb{D}\xi\cdot\mathbb{D}\eta}} = \frac{-\frac{1}{3}}{\sqrt{\frac{5}{6}\cdot\frac{1}{3}}} = - \sqrt{\frac{2}{5}}$
<br>Находим коэффициент $C = \frac{1}{(2\pi)^{\frac{n}{2}} \sqrt{det\Sigma}} = \frac{1}{2\pi\sqrt{6}}$

В итоге, плотность имеет следующий вид: $$p_{\xi,\eta}(x,y) = \frac{1}{2\pi\sqrt{6}}\exp \left(-\frac{1}{2}\left(
  \begin{pmatrix} x \\ y \end{pmatrix} - 
  \begin{pmatrix} 1 \\ 1\end{pmatrix}\right)^T 
  \begin{pmatrix} 2 & 2 \\ 2 & 5 \end{pmatrix} \left(
  \begin{pmatrix} x \\ y \end{pmatrix} - 
  \begin{pmatrix} 1 \\ 1\end{pmatrix}\right)\right)$$

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

C = 1 / (2 * np.pi * np.sqrt(6))
mean = np.array([1, 1], dtype=float)
sigma = 1 / 6 * np.array([[5, -2], [-2, 2]], dtype=float)

x = np.linspace(mean[0] - 2, mean[0] + 2, 300)
y = np.linspace(mean[1] - 2, mean[1] + 2, 300)
X, Y = np.meshgrid(x, y)

def f(x, y, sigma, mean, C):
    sigma_det = np.linalg.det(sigma)
    sigma_inv = np.linalg.inv(sigma)
    vec = np.array([x, y], dtype=float)
    return C * np.exp(-1/2 * (vec - mean).T.dot(sigma_inv).dot(vec - mean))

def plot_density(X, Y, Z, top_pov=False):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(X, Y, Z, cmap='viridis')
    if top_pov:
        ax.view_init(elev=90, azim=-90, roll=0)
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('p(x, y)')
    plt.title('График функции плотности вероятности')

    plt.show()

Z = np.zeros(X.shape)
for i in range(X.shape[0]):
    for j in range(X.shape[1]):
        Z[i, j] = f(X[i, j], Y[i, j], sigma, mean, C)

plot_density(X, Y, Z)
```

**5.** Найти условное распределение $\xi$ при условии $\eta$. Вычислить $\mathbb{E}\left(\xi|\eta\right)$ и $\mathbb{D}\left(\xi|\eta\right)$.

Для находения условного распределения воспользуемся формулой: $p_{\xi|\eta=y}(x) = \frac{p_{\xi,\eta}(x,y)}{p_{\eta}(y)}$.
