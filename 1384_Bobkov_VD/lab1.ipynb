## Основные теоретические положения
Случайный вектор $\overrightarrow{\xi} = \left(\xi_1, \dots, \xi_n\right)^\intercal:\Omega\to \mathbb{R}^n$ имеет многомерное нормальное распределение, если существует вектор независимых стандартных нормальных случайных величин $\overrightarrow{\eta} = \left(\eta_1, \dots, \eta_m\right)^\intercal$, матрица A размерности $n \times m$ и вещественный вектор $\overrightarrow{b} = (b_1,\dots,b_n)^\intercal$, такие что: 

 $\overrightarrow{\xi} = A\overrightarrow{\eta} + \overrightarrow{b} $.

Плотность многомерного невырожденного нормального распределения имеет вид: 
$p_{\overrightarrow{\xi}}\left(x_1, \dots, x_n\right) = \frac{1}{(2\pi)^{n/2}\sqrt{\det \Sigma}} \exp\left(-\frac{1}{2}\left(\overrightarrow{x}-\overrightarrow{\mu}\right)^\intercal \Sigma^{-1}\left(\overrightarrow{x}-\overrightarrow{\mu}\right)\right)$

где $\overrightarrow{\mu} = \mathbb{E}\overrightarrow{\xi} = \left(\mathbb{E}\xi_1, \dots, \mathbb{E}\xi_n\right)^\intercal$, $\Sigma = \left\{\sigma_{i,j}\right\}_{i,j=1}^n$, где $\sigma_{i,j} = \mathop{\mathrm{cov}}\left(\xi_i,\xi_j\right)$.

Свойства многомерного нормального распределения:
1. Линейная комбинация случайных переменных, принадлежащих многомерному нормальному распределению, также является нормально распределенной;
2. Условные и безусловные распределения, полученные из многомерного нормального распределения, также являются нормально распределенными;
3. Параметры многомерного нормального распределения полностью описывают его форму;
4. Многомерное нормальное распределение устойчиво относительно линейных преобразований. Если $\overrightarrow{\xi} \sim \mathcal{N}\left(\overrightarrow{\mu}, \Sigma\right)$, $A$ - произвольная матрица размерности $m \times n$, $\overrightarrow{b}$ - произвольный вектор размерности $m \times 1$, то $A\overrightarrow{\xi}+\overrightarrow{b} \sim \mathcal{N}\left(A\overrightarrow{\mu}+\overrightarrow{b}, A \Sigma A^\intercal\right)$.

### Задание 1
Вычислить вектор математических ожиданий и ковариационнные характеристики данного случайного вектора. Построить график плотности распределения $p_{\xi, \eta}$.

Используя метод Лагранжа, можно привести квадратичную форму к каноническому виду.

Пусть $q(x,y) = 7x^2 +8xy + 13y^2 + 6x - 18y +12 $

$q(x,y) = \left(7x^2 +8xy+6x\right) + 13y^2 - 18y + 12  = \left(7x^2 + 2x(4y + 3)\right) + 13y^2 - 18y + 12
= 7\cdot \left ( x^2 + 2x\left ( \frac{4}{7}y + \frac{3}{7} \right ) + \left ( \frac{1}{7}\left ( 4y + 3 \right ) \right ) ^ 2 \right ) - \frac{1}{7}\left ( 4y + 3 \right ) ^ 2 + 13y^2 - 18y + 12 = 7 \cdot \left ( x + \frac{1}{7} \left ( 4y + 3 \right ) \right ) ^ 2 + \frac{15}{7} \cdot \left ( 5y^2 - 10y + 5 \right ) = 7 \cdot \left ( x + \frac{1}{7} \left ( 4y + 3 \right ) \right ) ^ 2 + \frac{75}{7} \cdot \left ( y - 1 \right ) ^ 2 = 7 \cdot \left ( x + 1 + \frac{4}{7} \left ( y - 1 \right ) \right ) ^ 2 + \frac{75}{7} \cdot \left ( y - 1 \right ) ^ 2 = 7\left ( x + 1 \right ) ^ 2 + 8\left ( x + 1 \right )\left ( y - 1 \right ) + \left ( \frac{16}{7} + \frac{75}{7} \right )\left ( y - 1 \right ) ^ 2 = 7\left ( x + 1 \right ) ^ 2 + 8\left ( x + 1 \right )\left ( y - 1 \right ) + 13 \left ( y - 1 \right ) ^ 2 $

Теперь можно найти вектор математических ожиданий и обратную ковариационную матрицу:

$\mathbb{E} \begin{pmatrix} \xi \\ \eta \end{pmatrix} = \begin{pmatrix} -1 \\ 1 \end{pmatrix}$

$\Sigma^{-1} = \begin{pmatrix} 7 & 4 \\ 4 & 13 \end{pmatrix}$

В таком случае матрица ковариаций равна 

$\Sigma = \frac{1}{75} \begin{pmatrix} 13 & -4 \\ -4 & 7 \end{pmatrix}$

Дисперсии для каждой случайной величины равны 

$\mathbb{D}\xi = \frac{13}{75} \;\;\;\;\;\; \mathbb{D}\eta = \frac{7}{75}$

Нахождение ковариацию и коэффициент корреляции

$cov(\xi, \eta) = -\frac{4}{75}$

$\rho(\xi, \eta) = \frac{-\frac{4}{75}}{\frac{\sqrt{13 \cdot 7}}{75}} = -\frac{4}{\sqrt{91}}$

Вычисление C:

$C = \frac{1}{2\pi \sqrt{\frac{1}{75}}} = \frac{5 \sqrt{3}}{2\pi}$

Функция плотности распредления имеет вид:


$p_{\xi, \eta}(x,y) = \frac{5 \sqrt{3}}{2\pi} \cdot \exp \left(-\frac{1}{2} \begin{bmatrix}\begin{pmatrix} x \\ y \end{pmatrix}  - \begin{pmatrix} -1 \\ 1 \end{pmatrix} \end{bmatrix}^{\intercal} \cdot \begin{pmatrix} 7 & 4 \\ 4 & 13 \end{pmatrix} \cdot \begin{bmatrix}\begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} -1 \\ 1 \end{pmatrix} \end{bmatrix} \right)$

График плотности распределения $p_{\xi, \eta}$ построен ниже

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def f(x,y):
    return (5 * np.sqrt(3) / (np.pi * 2)) * np.exp(-0.5 * (7 * x ** 2 +8 * x * y + 13 * y ** 2 + 6 * x - 18 * y + 12))

x = np.linspace(-3, 1, 200)
y = np.linspace(-1, 2, 200)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
Graph_3D = fig.add_subplot(projection='3d')
Graph_3D.plot_surface(X,Y,Z,cmap='hot')
Graph_3D.set_xlabel('X')
Graph_3D.set_ylabel('Y')
Graph_3D.set_zlabel('Значение плотности распределения')
Graph_3D.set_title('График плотности распределения')
plt.show()
```

### Задание 3
Найти ортогональное преобразование, переводящее соответствующий центрированный случайный вектор в вектор с независимыми компонентами. Построить график плотности полученного распределения.

нужно найти собственные векторы матрицы $\Sigma^{-1}$

$\Sigma^{-1} = \begin{pmatrix} 7 & 4 \\ 4 & 13 \end{pmatrix}$

Для нахождения собсвтенных векторов нужно найти собестенные числа данной матрицы. Нахождение собственных чисел данной матрицы представлено ниже:

$ \begin{vmatrix} 7 - \lambda & 4 \\ 4 & 13 - \lambda \end{vmatrix} = 0$

$ \lambda^2 -20\lambda + 75 = 0$

$ \lambda_{1} = 5 \; \;  \lambda_{2} = 75$

$ \lambda_{1} = 5$ :

$\begin{pmatrix} 2 & 4 \\ 4 & 8\end{pmatrix} \begin{pmatrix} \upsilon_{1x} \\ \upsilon_{1y} \end{pmatrix} = \begin{pmatrix} 0 \\ 0 \end{pmatrix}$

Таким образом, $\upsilon_{1} =\frac{1}{\sqrt{5}} \begin{pmatrix} -2 \\ 1 \end{pmatrix}$

$ \lambda_{2} = 15$ :

$\begin{pmatrix} -8 & 4 \\ 4 & -2\end{pmatrix} \begin{pmatrix} \upsilon_{2x} \\ \upsilon_{2y} \end{pmatrix} = \begin{pmatrix} 0 \\ 0 \end{pmatrix}$

Таким образом, $\upsilon_{2} =\frac{1}{\sqrt{5}} \begin{pmatrix} 1 \\ 2 \end{pmatrix}$

Полученные собственные векторы являются нормированными.

Объединив векторы и транспонировав матрицу, получается матрица ортогонального преобразования.

$ Q^{\intercal} = \frac{1}{\sqrt{5}} \begin{pmatrix} -2 & 1 \\ 1 & 2 \end{pmatrix}$

$ Q = \frac{1}{\sqrt{5}} \begin{pmatrix} -2 & 1 \\ 1 & 2 \end{pmatrix}$

$\begin{pmatrix} \xi_{орт} \\ \eta_{орт} \end{pmatrix} = Q \begin{pmatrix} \xi \\ \eta \end{pmatrix} = \frac{1}{\sqrt{5}} \begin{pmatrix} -2\xi + \eta \\ \xi + 2\eta \end{pmatrix}$

Далее можно получить вектор математических ожиданий и матрицу ковариаций для $\xi_{орт} $ и $\eta_{орт}$ 

$\mathbb{E} \begin{pmatrix} \xi_{орт} \\ \eta_{орт} \end{pmatrix} = \frac{1}{\sqrt{5}} \begin{pmatrix} 2 + 1 \\ -1 + 2 \end{pmatrix} = \frac{1}{\sqrt{5}} \begin{pmatrix} 3 \\ 1 \end{pmatrix}$

$\Sigma_{ort} = \frac{1}{75} \cdot \frac{1}{5} \begin{pmatrix} -2 & 1 \\ 1 & 2 \end{pmatrix}  \begin{pmatrix} 13 & -4 \\ -4 & 7 \end{pmatrix} \begin{pmatrix} -2 & 1 \\ 1 & 2 \end{pmatrix} =  \begin{pmatrix}  \frac{1}{5} & 0 \\ 0 &  \frac{1}{15} \end{pmatrix} $

$\Sigma^{-1}_{ort} = \begin{pmatrix} 5 & 0 \\ 0 & 15 \end{pmatrix}$

Значения дисперсии для $\xi_{орт} $ и $\eta_{орт}$ указаны ниже:

$\mathbb{D}\xi_{ort} = \frac{1}{5} \;\;\;\;\;\; \mathbb{D}\eta_{ort} = \frac{1}{15} $

$cov(\xi_{орт}, \eta_{орт}) = 0 \;\; \rho(\xi_{орт}, \eta_{орт}) = 0$

Данные случайные величины являются независимыми, что подтверждается результатами, полученными выше.

Затем можно вычислить коэффициент C:

$C = \frac{1}{2\pi \sqrt{\frac{1}{75}}} = \frac{5 \sqrt{3}}{2\pi}$

Функция плотности распредления имеет вид:

$p_{\xi_{орт}, \eta_{орт}}(x,y) = \frac{5 \sqrt{3}}{2\pi} \cdot \exp \left(-\frac{1}{2} \begin{bmatrix} \begin{pmatrix} x \\ y \end{pmatrix}  - \begin{pmatrix} \frac{3}{\sqrt{5}} \\ \frac{1}{\sqrt{5}} \end{pmatrix} \end{bmatrix}^{\intercal} \cdot \begin{pmatrix} 5 & 0 \\ 0 & 15 \end{pmatrix} \cdot \begin{bmatrix}\begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} \frac{3}{\sqrt{5}} \\ \frac{1}{\sqrt{5}} \end{pmatrix} \end{bmatrix} \right)$



```python
def f(x,y):
    return (5 * np.sqrt(3) / (np.pi * 2)) * np.exp(-0.5 * (5*x**2 - 6*np.sqrt(5) * x +12 + 15*y**2 - 6*np.sqrt(5)*y ))

x = np.linspace(-1, 3, 200)
y = np.linspace(-1, 3, 200)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
Graph_3D = fig.add_subplot(projection='3d')
Graph_3D.plot_surface(X,Y,Z,cmap='hot')
Graph_3D.set_xlabel('X')
Graph_3D.set_ylabel('Y')
Graph_3D.set_zlabel('Значение плотности распределения')
Graph_3D.set_title('График плотности распределения')
plt.show()
```

$A = \begin{pmatrix} 1 & -4 \\ 2 & 2 \end{pmatrix}$

Нужно проверить, является ли данная матрица вырожденной:

$\det A = 10 $ 

Матрица A невырожденная, значит можно вычислить характеристики распределения:

$\begin{pmatrix} \xi^{'} \\ \eta^{'} \end{pmatrix} = \begin{pmatrix} 1 & -4 \\ 2 & 2 \end{pmatrix}\begin{pmatrix} \xi \\ \eta \end{pmatrix}$

$\Sigma^{'} = \frac{1}{75} \begin{pmatrix} 1 & -4 \\ 2 & 2 \end{pmatrix} \begin{pmatrix} 13 & -4 \\ -4 & 7 \end{pmatrix} \begin{pmatrix} 1 & 2 \\ -4 & 2 \end{pmatrix} = \begin{pmatrix} \frac{157}{75} & -\frac{6}{75} \\ -\frac{6}{75} & \frac{48}{75} \end{pmatrix}$

$\Sigma^{'  -1} = \begin{pmatrix} \frac{12}{25} & -\frac{3}{50} \\ -\frac{3}{50} & \frac{157}{100} \end{pmatrix}$

$C^{'} = \frac{1}{2\pi \sqrt{\frac{4}{3}}} = \frac{\sqrt{3}}{4\pi}$

$p_{\xi^{'}, \eta^{'}}(x,y) = \frac{\sqrt{3}}{4\pi} \cdot \exp \left(-\frac{1}{2} \begin{bmatrix}\begin{pmatrix} x \\ y \end{pmatrix}  - \begin{pmatrix} -5 \\ 0 \end{pmatrix} \end{bmatrix}^{\intercal} \cdot \begin{pmatrix} \frac{12}{25} & -\frac{3}{50} \\ -\frac{3}{50} & \frac{157}{100} \end{pmatrix} \cdot \begin{bmatrix}\begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} -5 \\ 0 \end{pmatrix} \end{bmatrix} \right)$
