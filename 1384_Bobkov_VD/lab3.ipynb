Линейная регрессия некоторой зависимой переменной y на набор независимых переменных  $x = (x_1, …, x_n)$, где n – это число наблюдений, предполагает, что линейное отношение между y и x: $y =\beta_0  + \beta_1 x_1 + ⋯ + \beta_n x_n + \varepsilon$ - уравнение регрессии. $\beta_0, \beta_1, …, \beta_n$ – коэффициенты регрессии (параметры), и $\varepsilon$ – случайная ошибка.

2. МНК (Метод Наименьших Квадратов)

$RSS(\beta)=\left\| Y-X^{\intercal}\beta \right\|=(Y-X^{\intercal}\beta)^{\intercal}\cdot(Y-X^{\intercal}\beta)\to min\beta$

Нормальные уравнения: $XX^{\intercal}\beta=XY$

Решение системы нормальных уравнений в регулярном случае: $\widehat{\beta}=(XX^{\intercal})^{-1}XY$

$RSS(\beta)=\sum_{i=1}^{n}\left( Y_{i}-\sum_{j=1}^{m}x_{ji}\beta_{j} \right)^{2}$

$\frac{\partial RSS}{\partial \beta_{k}} = -2\sum_{i=1}^{n}x_{ki}\left(Y_{i}- \sum_{j=1}^{m}x_{ji}\beta_{j}\right)=0, k=\overline{1,m}$

3. Несмещённая оценка дисперсии $(S^{2})$

$SS_{e}=RSS(\widehat{\beta})$, а r-ранг матрицы X

X на $N$ интервалов $I_{1}, ..., I_{N}:$ $I_{i} = (a_{i-1}, a_i], i = \overline {1,N}$. данные интервалы не пересекаются и покрывают всё множество возможных значений сл. вел. X. $p_{i} = P(X \in I_{i}|H_{0})$

Через $n_{i}$ обозначим число наблюдений нашей выборки, попавших в соответветствующий интервал $I_{i}$.

Статистика критерия основана на сравнении относительной частоты и теоритической.

 ### 1. Построить графически результаты эксперимента. Сформулировать линейную регрессионную модель переменной Y по переменной X. Построить МНК оценки параметром сдвига $\beta_0$ и масштаба $\beta_1$. Построить полученную линию регрессии. Оценить визуально соответствие полученных данных и построенной оценки.

Линейная регрессионная модель:

$ Y = X^T \beta + \varepsilon , \quad  \varepsilon $ - вектор ошибок 

$ E_\theta \varepsilon = 0 , \quad Var_\theta \varepsilon = \sigma^2 E $

$ E(Y|X) = \beta_0 + \beta_1X, \quad \beta = (\beta_0, \beta_1) $

МНК оценка:

$ \widehat \beta = (XX^T)^{-1}XY $

```python
beta_MNK = np.linalg.inv(X @ X.T) @ X @ Y
print("МНК оценка бета", beta_MNK)

model_values = X.T @ beta_MNK
plt.scatter(X[1,:], Y)
plt.plot(X[1,:],model_values,color = 'r')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Результаты эксперимента')
plt.show()
```

### 2. Построить и интерпретировать несмещённую оценку дисперсии. На базе ошибок построить гистограмму с шагом $h$. Проверить гипотезу нормальности ошибок на уровне значимости $\alpha$ по $\chi^2$. Оценить расстояние полученной оценки до класса нормальных распределений по Колмогорову. Визуально оценить данный факт.

Несмещенная оценка дисперсии

$ \sigma ^2 = \frac{1}{n-r} \cdot  (Y - X^T \widehat \beta)^T (Y - X^T \widehat \beta) $

$ r = 2$ - ранг X

```python
n = len(Y)
sigma_sq = (1/ (n - 2)) * (Y - X.T @ beta_MNK).T @ (Y - X.T @ beta_MNK)
print("Значение несмещенной sigma^2: ", sigma_sq)
print("Вектор ошибок")
miss_vector = Y - X.T @ beta_MNK
print(miss_vector)
hist, bins, _ = plt.hist(miss_vector, bins=np.arange(min(miss_vector), max(miss_vector) + h, h), edgecolor='black')
plt.title('Гистограмма ошибок')
plt.show()
```

Проверка гипотезы нормальности ошибок на уровне значимости $\alpha$ по $\chi^2$

$ H_0 : \varepsilon_1, ..., \varepsilon_n \sim \mathcal{N}(0, \sigma^2)$

Интервалы значений,содержащие не менее 5 элементов из вектора ошибок

$I_1 = (-\infty, -4.8489168], \; n_1 = 9$

$I_2 = (-4.8489168, -2.6489168], \; n_2 = 7$

$I_3 = (-2.6489168, -0.4489168], \; n_3 = 10$

$I_{4} = (-0.4489168, 1.7510832], \; n_{4} = 7$

$I_{5} = (1.7510832, 6.1510832], \; n_{5} = 8$

$I_{6} = (6.1510832, +\infty), \; n_{6} = 9$

```python
def chi2(sigma):
    intervals = np.array([-4.8489168,-2.6489168,-0.4489168,1.7510832,6.1510832])
    p = np.zeros(len(intervals) + 1)
    theor = np.array([9,7,10,7,8,9])
    p[0] = scp.norm.cdf(intervals[0],0,sigma)
    #print('hello')
    for i in range(1, len(intervals)):
        p[i] = round(scp.norm.cdf(intervals[i],0,sigma) - scp.norm.cdf(intervals[i-1],0,sigma),10)
    p[len(intervals)] = 1 - scp.norm.cdf(intervals[-1],0,sigma)
    krit = ((theor - n * p) ** 2)/ (n * p)
    return np.sum(krit)
    
result = scipy.optimize.minimize_scalar(chi2)

print("Значение критерия Хи-квадрат: ", result.fun)

func = scp.chi2(4)
x_a = func.ppf(1 - alpha)

print("Квантиль: ", x_a)

print("Наибольшее значение уровня значимости, при котором еще нет оснований отвергнуть данную гипотезу",1- scp.chi2.cdf(result.fun, 6-1-1))
```

Полученное значение критерия $ \chi^2  = 2.982078005592144$

Теоретическое значение $ \chi^2$ для данного уровня доверия равно 11.667843403834778, следовательно, гипотезу можно принять.

Наибольшее значение уровня значимости, на котором нет оснований отвергнуть гипотезу, равно 0.5608290693214468

Оценка расстояния полученной оценки до класса нормальных распределений по Колмогорову.

$ H_0 : \varepsilon_1, ..., \varepsilon_n \sim \mathcal{N}(0, \widehat {\sigma^2}) $

$ \sqrt{n} \cdot max|F_n(y) - F(y)| \sim K $, $K$ - распределение Колмогорова. 

```python

D = scp.kstest([element for element in miss_vector], "norm", args = (0,np.sqrt(sigma_sq))).statistic
print("Статистика критерия Колмогорова Dn:", D)
print("x_alpha:", scp.kstwobign.ppf(1 - alpha) / np.sqrt(len(miss_vector)))
print("Наибольшее значение уровня значимости, при котором нет еще оснований отвергнуть данную гипотезу",1 - scp.ksone.cdf(D,n) )
```

$I_1 = (-\infty, -4.82534475], \; n_1 = 9$

$I_2 = (-4.82534475, -2.62534475], \; n_2 = 7$

$I_3 = (-2.62534475, -0.42534475], \; n_3 = 10$

$I_{4} = (-0.42534475, 1.77465525], \; n_{4} = 7$

$I_{5} = (1.77465525, 6.17465525], \; n_{5} = 8$

$I_{6} = (6.1510832, +\infty), \; n_{6} = 9$

```python
def chi2(sigma):
    intervals = np.array([-4.82534475, -2.62534475, -0.42534475, 1.77465525,6.17465525])
    p = np.zeros(len(intervals) + 1)
    theor = np.array([9,7,10,7,8,9])
    p[0] = scp.norm.cdf(intervals[0],0,sigma)
    for i in range(1, len(intervals)):
        p[i] = round(scp.norm.cdf(intervals[i],0,sigma) - scp.norm.cdf(intervals[i-1],0,sigma),10)
    p[len(intervals)] = 1 - scp.norm.cdf(intervals[-1],0,sigma)
    krit = ((theor - n * p) ** 2)/ (n * p)
    return np.sum(krit)
    
result = scipy.optimize.minimize_scalar(chi2)

print("Значение критерия Хи-квадрат: ", result.fun)

func = scp.chi2(4)
x_a = func.ppf(1 - alpha)

print("Квантиль: ", x_a)

print("Наибольшее значение уровня значимости, при котором еще нет оснований отвергнуть данную гипотезу",1- scp.chi2.cdf(result.fun, 6-1-1))
```

### 8. Сформулировать гипотезу линейной регрессионной зависимости переменной $Y$ от переменной $X$ и проверить её значимость на уровне $\alpha$.

Основная гипотеза $ H_0 : \beta_2 = 0$

$\mathcal{F} = \cfrac{\widehat \psi^T ((C^T (XX^T)^{-1}C)^{-1}) \widehat \psi}{qs^2} \sim F_{q, n-r}$

$x_\alpha: F_{q, n-r}(x_\alpha) = 1 - \alpha $

$ C = \begin{pmatrix}
    0 \\
    0 \\
    1
\end{pmatrix} $, $\widehat \psi = \widehat \beta_2, q = 1, r = 3$

$\mathcal{F} = \cfrac{{\widehat \beta_2} ^2}{bqs^2} $
