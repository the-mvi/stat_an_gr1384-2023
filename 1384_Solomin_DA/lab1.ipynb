## Основные теоретические положения

Случайный вектор $\vec\zeta$ называют нормальным, если $\vec\zeta = A\vecη + \vec{B}$, где

$\vecη = (\eta_1, ..., \eta_m)^{\intercal}$ - вектор стандартных нормально распределенных с.в.

$A$ -- матрица $(n \times m)$

$B = (b_1, ..., b_n)^{\intercal}$ -- вещественный вектор

Обозначение:

$\vec{\zeta} \sim N(μ, Σ)$, где $\mu = M(\vec{\zeta})$

$\sigma = \{σ_{ij}\}^{n}_{ij=1}$, $\sigma = cov(\zeta_i, \zeta_j)$

Плотность нормального вектора

$\frac{1}{(2\pi)^{n/2}\sqrt{\det{Σ}}} \cdot exp\left(-1/2(\vec{X} - μ)^{\intercal}Σ^{-1}(\vec{X} - \mu)^{\intercal}\right)$

Свойства:

1. линейная комбинация многомерной нормально-распределенной случайном величины (МНРСВ) является МНРСВ
1. Условные и безусловные рампределения являются МНРСВ
1. Параметры МНРСВ полностью его описывают
1. $\vec\zeta = A\vecη + \vec{B} \sim N(Aμ + B, AΣA^{\intercal})$

## Выполнение работы

$p_{ɛ,η}(x, y) = C \cdot exp(-\frac{1}{2}(3x^2 + 3xy + 7y^2 + 15x - 5y + 25))$




**1.** Вычислить вектор математических ожиданий и ковариационные характеристики данного случайного вектора. Построить график плотности распределения $p_{ɛ,η}$

$s(x, y) = 3x^2 + 3xy + 7y^2 + 15x - 5y + 25 = $

$= (3x^2 + 3xy + 15x) + 7y^2 - 5y + 25 =$

$= (3x^2 + 3x(y + 5)) + 7y^2 - 5y + 25 =$

$= 3 (x^2 + x(y + 5) + \frac{1}{4}(y+5)^2) + 7y^2 - 5y + 25 - \frac{3}{4}(y+5)^2 =$

$= \frac{3}{4}(2x + y + 5)^2 +7y^2 - 5y + 25 - \frac{3y^2 + 30y + 75}{4} = $

$= \frac{3}{4}(2x + y + 5)^2 + \frac{25}{4}(y-1)^2 =$

$= \frac{3}{4}(2(x + 3) + (y - 1))^2 + \frac{25}{4}(y-1)^2 = $

$= 3(x + 3)^2 + 3(x + 3)(y - 1)  + \frac{3}{4}(y - 1)^2 + \frac{25}{4}(y-1)^2 = $

$= 3(x + 3)^2 + 3(x + 3)(y - 1)  + 7(y-1)^2 = $

Отсюда:
$\vec{\mu} = \left(\begin{array}{cc} -3 \\ 1 \end{array}\right)$  


$\Sigma^{-1} = \left(\begin{array}{cc} 3 && 1.5 \\ 1.5 && 7 \end{array}\right)$

$\Sigma = \frac{4}{75}\left(\begin{array}{cc} 7 && -1.5 \\ -1.5 && 3 \end{array}\right)$  

$\mathbb{D}(\zeta) = \frac{28}{75}$  

$\mathbb{D}(\eta) = \frac{12}{75}$  


$ρ_{\zeta, \eta} = \frac{cov(\zeta, \eta)}{\sigma_{\zeta}\sigma_{η}} = \frac{75\sqrt{21}}{56}$  


$C = \frac{1}{2\pi \cdot \sqrt{\det\Sigma}} = \frac{1}{5\pi\sqrt{3}} $

$p_{\zeta. \eta}(x,y) = \frac{1}{5\pi\sqrt{3}}\cdot exp\left(-\frac{1}{2}\left[\left(\begin{array}{cc} x \\ y \end{array}\right) - \left(\begin{array}{cc} -3 \\ 1 \end{array}\right)\right]^{\intercal} \left(\begin{array}{cc} 3 && 1.5 \\ 1.5 && 7 \end{array}\right) \times  \left[\left(\begin{array}{cc} x \\ y \end{array}\right) - \left(\begin{array}{cc} -3 \\ 1 \end{array}\right) \right]^{\intercal}\right)$

```python
import seaborn as sb
import matplotlib.pyplot as plot
import numpy as np
import numexpr as ne


sb.set_style('darkgrid')
Sigma = np.array([
    [7*(4/75), -1.5*(4/75)],
    [-1.5*(4/75), 3*(4/75)]
])

M_zeta = -3
M_eta = 1
def draw():

  iverseSigma = np.linalg.inv(Sigma)
  C = 1 / (2 * np.pi * np.sqrt(np.linalg.det(iverseSigma)))

  function = f'{iverseSigma[0, 0]}*(x + {-M_zeta}) ** 2 + {iverseSigma[1, 0] *2}*(x + {-M_zeta})*(y - {M_eta}) + {iverseSigma[1, 1]}*(y - {M_eta}) ** 2'

  print(function)

  sigma3_zeta = Sigma[0, 0] ** 0.5 * 3
  sigam3_eta = Sigma[1, 1] ** 0.5 * 3

  x_plot1 = np.linspace(M_zeta - sigma3_zeta, M_zeta + sigma3_zeta, 50)
  y_plot1 = np.linspace(M_eta - sigam3_eta, M_eta + sigam3_eta, 50)
  plot1, plot2 = np.meshgrid(x_plot1, y_plot1)
  plot3 = C * np.exp(-1/2 * ne.evaluate(function, {'x': plot1, 'y': plot2}))

  fig = plot.figure(figsize=(8,8))

  axes = fig.add_subplot(projection='3d')
  axes.plot_surface(plot1, plot2, plot3)
  plot.show()

draw()


```
