## Основные теоретические положения
Случайный вектор $\overrightarrow{\xi} = \left(\xi_1, \dots, \xi_n\right)^\intercal:\Omega\to \mathbb{R}^n$ имеет многомерное нормальное распределение, если существует вектор независимых стандартных нормальных случайных величин $\overrightarrow{\eta} = \left(\eta_1, \dots, \eta_m\right)^\intercal$, матрица $A$ размерности $n \times m$ и вещественный вектор $\overrightarrow{b} = (b_1,\dots,b_n)^\intercal$, такие что: $\overrightarrow{\xi} = A\overrightarrow{\eta} + \overrightarrow{b}$.

Плотность многомерного невырожденного нормального распределения имеет вид: $\mathcal{p}_{\overrightarrow{\xi}}\left(x_1, \dots, x_n\right) = \frac{1}{(2\pi)^{n/2}\sqrt{\det \Sigma}} \exp\left(-\frac{1}{2}\left(\overrightarrow{x}-\overrightarrow{\mu}\right)^\intercal \Sigma^{-1}\left(\overrightarrow{x}-\overrightarrow{\mu}\right)\right)$, 
где $\overrightarrow{\mu} = \mathbb{E}\overrightarrow{\xi} = \left(\mathbb{E}\xi_1, \dots, \mathbb{E}\xi_n\right)^\intercal$ - вектор математических ожиданий компонент, 
$\Sigma = \left\{\sigma_{i,j}\right\}_{i,j=1}^n$ - ковариационная матрица, $\sigma_{i,j} = \mathop{\mathrm{cov}}\left(\xi_i,\xi_j\right)$

Свойства многомерного нормального распределения:
- Линейная комбинация случайных переменных, принадлежащих многомерному нормальному распределению, также является нормально распределенной
- Условные и безусловные распределения, полученные из многомерного нормального распределения, также являются нормально распределенными.
- Параметры многомерного нормального распределения (средние и ковариационная матрица) полностью описывают его форму
- Многомерное нормальное распределение устойчиво относительно линейных преобразований. Если $\overrightarrow{\xi} \sim \mathcal{N}\left(\overrightarrow{\mu}, \Sigma\right)$, $A$ - произвольная матрица размерности $m \times n$, a $\overrightarrow{b}$ - произвольный вектор размерности $m \times 1$, то $A\overrightarrow{\xi}+\overrightarrow{b} \sim \mathcal{N}\left(A\overrightarrow{\mu}+\overrightarrow{b}, A \Sigma A^\intercal\right)$

### 5. Найти условное распределение $\xi$ при условии $\eta$. Вычислить $\mathbb{E}(\xi|\eta)$ и $\mathbb{D}(\xi|\eta)$.

Из пункта 1:

$\mathbb{E}_{\eta} = -1$

$\mathbb{D}_{\eta} = \frac{1}{3}$

Тогда:

$
p_{\eta}(y) = \frac{\sqrt{3}}{\sqrt{2\pi}} \cdot \exp(-\frac{1}{2}\cdot(y+1)\cdot3\cdot(y+1)) = \frac{\sqrt{3}}{\sqrt{2\pi}} \cdot \exp(-\frac{3}{2}(y+1)^{2})
$

Найдем плотность условного распределения:

$\begin{split}
p_{\xi|\eta=y}(x) &= \frac{p_{\xi,\eta}(x,y)}{p_{\eta}(y)} = \\
&= \frac{\frac{\sqrt{6}}{2\pi}\cdot\exp\left(-\frac{1}{2}(2x^{2} + 5y^{2} - 8x - 2y + 4xy +11)\right)}{\frac{\sqrt{3}}{\sqrt{2\pi}}\cdot\exp\left(-\frac{1}{2}(3y^{2} + 6y + 3)\right)} = \\
&= \frac{1}{\sqrt{\pi}} \cdot \exp\left(-\frac{1}{2}(2x^2 + 5y^{2} - 3y^{2} - 8x - 2y - 6y + 4xy + 11 -3)\right) = \\
&= \frac{1}{\sqrt{\pi}} \cdot \exp\left(-\frac{1}{2}(2x^2 + 2y^{2} - 8x - 8y + 4xy + 8)\right) = \\
&= \frac{1}{\sqrt{\pi}} \cdot \exp\left(-\frac{1}{2}(2x^{2} + 4x(y-2) +2(y-2)^{2}\right) = \\
&= \frac{1}{\sqrt{\pi}} \cdot \exp\left(-\frac{1}{2}(2x^{2} + 4x(y-2) +2(y-2)^{2}\right) = \\
&= \frac{1}{\sqrt{\pi}} \cdot \exp\left(-\frac{1}{2}(2(x + (y-2)^{2})\right)
\end{split}$

Тогда математическое ожидание и дисперсия:

$\mathbb{E}(\xi|\eta = y) = -(y-2)$

$\mathbb{D}(\xi|\eta = y) = \frac{1}{2}$
