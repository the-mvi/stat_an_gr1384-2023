# Лабораторная работа №1 "Исследование двумерного нормального распределения"


## Цель работы  
Исследовать свойства двумерного случайного вектора имеющего нормальное распределение, овладеть навыками преобразования нормального вектора в стандартный и в вектор с независимыми компонентами.  

## Основные теоретические положения
$\vec{\xi}=(\xi_{1}\ldots\xi_{n})^\top\Omega\rightarrow\mathbb{R}^n$ - нормальный (Гауссовский), если существует вектор независимых стандартных случайных величин $\vec{\eta} = (\eta_{1} \ldots \eta_{m})^{\top}$ матрица $A$ размерности $m\times n$ и вещественный вектор $\vec B=(b_1 ... b_n)^\top$ такие, что: $\vec{\xi}=A\vec{\eta}+B$   

Обозначение: $\vec{\xi}\sim\mathcal{N}(\mu, \sigma^2)$, где $\mu=\mathcal{E}\vec{\xi}=(\mathcal{E}\xi_{1} \ldots \mathcal{E}\xi_{n})^{\top}$ - вектор математических ожиданий компонент.
$\Sigma=(\sigma_{ij})_{i}^{n}$, $i=j$ - матрица ковариаций, где
$\sigma_{ij}=cov(\xi_{i}, \xi_{j})$
Плотность распределения нормального вектора:
$\rho_{\xi,\eta}(x_{1} \ldots x_{n})=\frac{\mathop{\mathrm{cov}}\nolimits(\xi, \eta)}{\sqrt[]{\mathbb{D}_\xi\cdot\mathbb{D}_\eta}}exp(\frac{1}{2}(\vec{x}-\mu)^{\top}\Sigma^{-1}(\vec{x}-\mu))$, где $\vec{x}=(x_{1} \ldots x_{n})^{\top}$

Свойства:
1) Линейная комбинация случайных переменных, принадлежащих многомерному нормальному распределению, также является нормально распределенной.
2) Условное и безусловное распределения (полученные из многомерного норамльного распределения) являются нормальным распределением.
3) Параметры многомерного нормального распределения полностью описывают его форму.
4) Многомерное нормальное распределение устойчиво относительно линейных преобразований: если $\vec{\xi}\sim\mathcal{N}(\mu, \sigma^2)$, $A$ - произвольная невырожденная матрица размерности $m\times n$, а $B$ - произвольный вектор размера $m\times1$, то $A\vec{\xi}+B\sim\mathcal{N}(A\mu+B, A\Sigma A^\top)$

## Постановка задачи
Плотность двумерного нормального разпределениея имеет вид:
$\rho_{\xi,\eta}(x,y)=C\cdot(exp(-\frac{1}{2}(3x^2+3xy+7y^2-9x+8y+13)))$
    
1. Вычислить вектор математических ожиданий и ковариационные зарактеристики данного случайного вектора. Построить график плотности распределения $\rho_{\xi,\eta}$
2. Найти афинное преобразование, переводящее исходный случайный вектор в стандартный нормальный.
3. Найти ортогональное преобразование, переводящее соответствующий центрированный случайный вектор в вектор с независимыми компонентами. Построить график плотности полученного распределения.

**5. Нахождение условного распределения $\xi$ при условии $\eta$. Вычисление математического ожидания и дисперсии.**

Для нахождения условного распределения воспользуемся формулой:$$\rho_{\xi|\eta=y}(x)=\frac{\rho_{\xi,\eta}(x,y)}{\rho_{\eta}(y)},$$ где $\rho_{\eta}(y)$ можно найти по формуле:
$$\int_{-\infty}^\infty\rho_{\xi,\eta}(x,y)dy$$

Давайте вычислим интеграл:\\

$\int_{-\infty}^\infty \frac{1}{2\pi\sqrt[]{23.25}}exp(-\frac{1}{2}(3x^2+3xy+7y^2-9x+8y+13))dy=\frac{1}{\sqrt[]{2\pi\cdot7\cdot23.25}}exp(-\frac{75}{56}(x-2)^2)$\\
 
Получили распределение случайной величины $\eta$. Теперь найдем условное распределение\\

$\rho_{\xi|\eta=y}(x)= \frac{\frac{1}{2\pi\sqrt[]{23.25}}exp(-\frac{1}{2}(3x^2+3xy+7y^2-9x+8y+13))}{\frac{1}{\sqrt[]{2\pi\cdot7\cdot23.25}}exp(-\frac{75}{56}(x-2)^2)}=\sqrt[]{\frac{7}{2\pi}}exp(-\frac{1}{56}(9x^2+84x+48x+196y^2+224y+64))$\\

Далее вычислим дисперсию и математисеское ожидание для полученного условного распределения:\\

$\mathbb{E}_{\xi|\eta=y}=\int_\mathbb{R}x\cdot\rho_{\xi|\eta=y}(x)dx=\int_{-\infty}^\infty x\cdot\sqrt[]{\frac{7}{2\pi}}exp(-\frac{1}{56}(9x^2+84x+48x+196y^2+224y+64))=-\frac{44}{9}\sqrt[]{14\pi} exp(-\frac{1}{2}(y-1)(7y+15))$\\

$\mathbb{D}_{\xi|\eta=y}=\int_\mathbb{R}x^2\cdot\rho_{\xi|\eta=y}(x)dx=\int_{-\infty}^\infty x^2\cdot\sqrt[]{\frac{7}{2\pi}}exp(-\frac{1}{56}(9x^2+84x+48x+196y^2+224y+64))=\frac{1024}{27}\sqrt[]{14\pi} exp(-\frac{1}{2}(y-1)(7y+15))$\\
