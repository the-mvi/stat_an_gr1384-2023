## Основные теоретические положения
**Определения**:
- Нормальный случайный вектор $\vec{\xi} = (\xi_1, \dots, \xi_n)^T$ : $\Omega → {\mathbb R}^n $  называют таковым, если существует вектор независимых стандартных случайных величин  $ \vec{\eta} = (\eta_1, \dots, \eta_m)^T$, матрица *A* с размерностью $n × m$ и вещественный вектор $\vec B = (b_1, \dots, b_n)$, такие что: $ \space \vec{\xi}=A \vec{\eta} + \vec{B} $
- Плотностью многомерного невырожденного нормального распределения будем называть $\rho_{\vec{\xi}}(x_1, \dots, x_n) = \frac 1 {(2 \pi)^{n/_2} \sqrt{det \Sigma}} \exp(-\frac 1{2} (\vec{x} - \vec{\mu})^T \Sigma^{-1} (\vec{x} - \vec{\mu}))$, где:
\
$ \vec{x} = (x_1, \dots, x_n)^T $, $ \vec{\xi} \sim \mathcal{N}(\vec{\mu}, \Sigma)$, где $ \vec{\mu} = \mathbb{E} \vec{\xi}$ и $\Sigma = \{cov(\xi_i, \xi_j)\}_{i = 1, j = 1}^n$

**Свойства:**
1. Линейная комбинация случайных переменных, пренадлежащих многомерному нормальному распределению, также является нормально распределённой.
2. Условные и безусловные распределения, полученные из многомерного нормального распределения, также являются нормальными распределениями.
3. Параметры многомерного нормального распределения польностью описывают его форму.
4. Многомерное нормальное распределение устойчиво относительно линейных преобразования \
Если $\vec{\xi} \sim \mathcal{N}(\vec{\mu}, \Sigma)$, $A$  $-$ произвольная матрица размерности $m×n$, $\space \vec B$ $-$ произвольный вектор размерности $m×1$, то $ A \vec{\xi} + \vec{B} \sim \mathcal{N}(A \vec{\mu} + \vec{B}, A \Sigma A^T) $

### 1 задание. Вычислить вектор математических ожиданий и ковариационные характеристики данного случайного вектора. Построить график плотности распределения $ \rho_{\xi, \eta} $.
Приведём квадратичную форму к каноническому виду. Пусть $q(x,y) = (6x^2 -4xy +9y^2 - 8x - 14y + 11)$, тогда:
$q(x,y) = 6(x^2 - \frac 2{3} xy - \frac 4{3}x) +9y^2  - 14y + 11 = 6(x^2 - 2x (\frac 1{3} y + \frac 2{3}) + (\frac 1{3} y + \frac 2{3})^2 - (\frac 1{3} y + \frac 2{3})^2) +9y^2  - 14y + 11 = 6((x - \frac 1{3} y - \frac 2{3})^2 - (\frac 2{3}y^2  + \frac 8{3}y + \frac 8{3})) +9y^2  - 14y + 11 = 6(x - \frac 1{3} y - \frac 2{3})^2 + \frac {25}{3}y^2  - \frac {50}{3}y + \frac {25}{3} = \mathbf {6(x - \frac 1{3} y - \frac 2{3})^2 + \frac {25}{3} (y-1)^2} = 6(-\frac13 (y-1) -1 + x)^2 + \frac {25}{3} (y-1)^2 = \mathbf {6(x-1)^2 -4(x-1)(y-1)+9(y-1)^2}$
\
\
Найдём вектор мат. ожиданий и обратную ковариационную матрицу:
$\mathbb{E}\begin{pmatrix}\xi \\\eta\end{pmatrix}=\begin{pmatrix}1 \\1\end{pmatrix}\quad\Sigma^{-1} =\begin{pmatrix}6 & -2 \\-2 & 9\end{pmatrix} \quad det \Sigma^{-1} = 54 - 4 = 50 \implies\Sigma = \frac{1}{50} \begin{pmatrix}9 & 2 \\ 2 & 6 \end{pmatrix} $
\
\
Найдём ковариационные характеристики данного случайного вектора:
$ cov(\xi, \eta) = \frac{1}{25} \quad \mathbb{D}_{\xi} = \frac{9}{50}  \quad \mathbb{D}_{\eta} = \frac{3}{25} \implies \rho(\xi, \eta)=\frac{cov({\xi},{ \eta})}{\sqrt{\mathbb{D}_{\xi} \mathbb{D}_{\eta}}} = \frac{\frac{1}{25}}{\sqrt{\frac{9}{50} \frac{3}{25}}} = \frac {\frac{1}{25}}{\frac{3}{\sqrt{50}}\frac{\sqrt3}{5}} =\frac{10}{3\sqrt{150}} = \frac{10\sqrt{150}}{3*150} = \frac{\sqrt6}{9}$
\
\
Найдём коэффициент плотности распределения $C$: \
$C = \frac{1}{(2 \pi)^{\frac{n}{2}} \sqrt{det \Sigma}} = \frac{1}{2 \pi \sqrt{\frac{1}{50}}} = \frac{5 \sqrt{2}}{2 \pi}=\frac{5}{\pi \sqrt{2}}$
\
\
Тогда функция плотности будет выглядеть следующим образом:
$\rho_{\xi, \eta}(x, y)= \frac{5}{\pi \sqrt{2}} \cdot \exp(-\frac{1}{2}(6x^2+4xy+9y^2+8x-14y+11)) $



```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

def p(x, y):
    return (5 / (np.pi * np.sqrt(2))) * np.exp(-0.5 * (6 * x ** 2 - 4 * x * y + 9 * y ** 2 - 8 * x - 14 * y + 11))

x = np.linspace(-1, 3, 200)
y = np.linspace(-1, 3, 200)

X, Y = np.meshgrid(x, y)
Z = p(X, Y)

fig = plt.figure(figsize=(10,8))
ax = plt.axes(projection='3d')

ax.plot_surface(X, Y, Z, cmap='rainbow')
ax.set_title('Density graph')
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('y', fontsize=12)
ax.set_zlabel('z', fontsize=12)
plt.show()
```

### 3 задание. Найти ортогональное преобразование, переводящее соответствующий центрированный случайный вектор в вектор с независимыми компонентами. Построить график плотности полученного распределения.
$\Sigma^{-1} = \begin{pmatrix} 6 & -2 \\ -2 & 9 \end{pmatrix}$

Для получение ортогонального преобразования найдём собственные числа матрицы $\Sigma^{-1}$ с помощью уравнения $\lambda^2 - th( \Sigma^{-1} ) + det( \Sigma^{-1} ) = 0$. \
$ th(\Sigma^{-1}) = 15 \quad det(\Sigma^{-1}) = 50 \implies \lambda^2 - 15 \lambda + 50 = 0$
\
$\lambda_{1} = 5 \space\space \lambda_{2} = 10$
\
\
При $\lambda = 5$:
\
$\begin{pmatrix} 1 & -2 \\ -2 & 4 \end{pmatrix}
\begin{pmatrix} \upsilon_1^{(1)} \\ \upsilon_2^{(1)}\end{pmatrix} = 0
\space ⇒
\begin{pmatrix} 2 \\ 1 \end{pmatrix}\frac{1}{\sqrt5} = \upsilon^{(1)} \space - $ нормированный собственный вектор.
\
\
При $\lambda = 10$:
\
$\begin{pmatrix} -4 & -2 \\ -2 & -1 \end{pmatrix} \begin{pmatrix} \upsilon_1^{(2)} \\ \upsilon_2^{(2)}\end{pmatrix} = 0 \space ⇒ \begin{pmatrix} -1 \\ 2 \end{pmatrix}\frac{1}{\sqrt5} = \upsilon^{(2)} \space - $ нормированный собственный вектор.
\
\
Объединив векторы и транспонировав матрицу, получим матрицу ортогонального преобразования:
$ Q^{\intercal} = \frac{1}{\sqrt{5}} \begin{pmatrix} 2 & -1 \\ 1 & 2 \end{pmatrix} \space Q = \frac{1}{\sqrt{5}} \begin{pmatrix} 2 & 1 \\ -1 & 2 \end{pmatrix}$
\
\
Найдём $E \vec{\xi_{ort}} $ и $\Sigma_{ort} $:

$E \begin{pmatrix} \xi_{ort} \\ \eta_{ort} \end{pmatrix}  = \frac{1}{\sqrt{5}}\begin{pmatrix} 2 & 1 \\ -1 & 2 \end{pmatrix}\begin{pmatrix} 1 \\ 1 \end{pmatrix} = \frac{1}{\sqrt{5}}  \begin{pmatrix} 3 \\ 1 \end{pmatrix}$

$ \Sigma_{ort}=Q \Sigma Q^T = \frac{1}{50}\frac{1}{5} \begin{pmatrix} 2 & 1 \\ -1 & 2 \end{pmatrix}\begin{pmatrix} 9 & 2 \\ 2 & 6 \end{pmatrix}\begin{pmatrix} 2 & -1 \\ 1 & 2 \end{pmatrix} =  \frac{1}{250}\begin{pmatrix} 20 & 10 \\ -5 & 10 \end{pmatrix}\begin{pmatrix} 2 & -1 \\ 1 & 2 \end{pmatrix} = \frac{1}{250}\begin{pmatrix} 50 & 0 \\ 0 & 25 \end{pmatrix} =\begin{pmatrix} \frac{1}{5} & 0 \\ 0 & \frac{1}{10} \end{pmatrix}$
\
\
$\Sigma^{-1}_{ort} = \begin{pmatrix} 5 & 0 \\ 0 & 10 \end{pmatrix}$

$D\xi_{ort} = \frac{1}{5} \space \space \space D\eta_{ort} = \frac{1}{10} $

$cov(\xi_{ort}, \eta_{ort}) = 0 \;\; \rho(\xi_{ort}, \eta_{ort}) = 0$

$C = \frac{1}{2\pi \sqrt{\frac{1}{50}}} = \frac{5}{\pi\sqrt{2}}$

$p_{\xi_{ort}, \eta_{ort}}(x,y) = \frac{5}{\pi\sqrt{2}} * \exp \left(-\frac{1}{2} \begin{bmatrix} \begin{pmatrix} x \\ y \end{pmatrix}  - \begin{pmatrix} \frac{3}{\sqrt{5}} \\ \frac{1}{\sqrt{5}} \end{pmatrix} \end{bmatrix}^{\intercal} * \begin{pmatrix} 5 & 0 \\ 0 & 10 \end{pmatrix} * \begin{bmatrix}\begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} \frac{3}{\sqrt{5}} \\ \frac{1}{\sqrt{5}} \end{pmatrix} \end{bmatrix} \right)$

```python
def p(x, y):
    return (5 / (np.pi * np.sqrt(2))) * np.exp(-0.5 * (5*x**2 + (-30/np.sqrt(5))*x + 11 + 10*y**2 + (-20/np.sqrt(5))*y))


x = np.linspace(-1, 3, 200)
y = np.linspace(-2, 3, 200)

X, Y = np.meshgrid(x, y)
Z = p(X, Y)

fig = plt.figure(figsize=(10,8))
ax = plt.axes(projection='3d')

ax.plot_surface(X, Y, Z, cmap='rainbow')
ax.set_title('График плотности')
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('y', fontsize=12)
ax.set_zlabel('z', fontsize=12)
plt.show()
```

### Задание 4. Вычислить характеристики совместного распределения случайного вектора $ (-\xi + 2 \eta, 3 \xi - 3 \eta)^T $ и записать его плотность. Построить график плотности полученного распределения.

Представим новый вектор в виде:
$ \vec{\xi_н} = A \vec{\xi} + \vec{B}\space$:
\
$ \vec{\xi_н} = \begin{pmatrix} - \xi + 2 \eta \\ 3 \xi - 3 \eta \end{pmatrix} = \begin{pmatrix} -1 & 2 \\ 3 &-3 \end{pmatrix} \begin{pmatrix}\xi \\ \eta \end{pmatrix} + \begin{pmatrix} 0 \\ 0 \end{pmatrix}$
\
\
Найдём $E \vec{\xi_{н}} $ и $\Sigma_{н} $. \
$ E\begin{pmatrix} \xi_{н} \\ \eta_{н} \end{pmatrix}  = \begin{pmatrix} -1 & 2  \\ 3 & -3 \end{pmatrix} \begin{pmatrix} 1 \\ 1 \end{pmatrix} = \begin{pmatrix} 1 \\ 0 \end{pmatrix}$
\
\
$ \Sigma_{н}=A \Sigma A^T = \frac{1}{50} \begin{pmatrix} -1 & 2 \\ 3 & -3 \end{pmatrix} \begin{pmatrix} 9 & 2 \\ 2 & 6 \end{pmatrix} \begin{pmatrix} -1 & 3 \\ 2 & -3 \end{pmatrix} =  \frac{1}{50} \begin{pmatrix} -5 & 10 \\ 21 & -12 \end{pmatrix} \begin{pmatrix} -1 & 3 \\ 2 & -3 \end{pmatrix} = \frac{1}{50} \begin{pmatrix} 25 & -45 \\ -45 & 99 \end{pmatrix} $
\
\
$det \Sigma = \frac{99 * 25-45^2}{50^2} = \frac{9}{50} \implies \Sigma_н^{-1} = \frac{1}{9}
\begin{pmatrix} 99 & 45 \\ 45 & 25 \end{pmatrix}$
\
\
Найдём характеристики совместного распределения:
$ cov(\xi_н, \eta_н) = -\frac{9}{10} \quad D_{\xi_н} = \frac12 \quad D_{\eta_н} = \frac{99}{50} \implies \rho(\xi_н, \eta_н)=\frac{cov({\xi_н},{ \eta_н})}{\sqrt{D_{\xi} D_{\eta}}} = \frac{-\frac{9}{10}}{\sqrt{\frac{99}{100}}} = \frac{-\frac{9}{10}}{\frac{\sqrt{99}}{10}} = -\frac{3\sqrt{11}}{11}$
\
\
Найдём коэффициент плотности распределения $C$. \
$C = \frac{1}{(2 \pi)^{\frac{n}{2}} \sqrt{det \Sigma_н}} = \frac{1}{2 \pi \frac{3}{5\sqrt{2}}} = \frac{5\sqrt2}{6\pi}$ \
\
$p_{\xi_{4}, \eta_{4}}(x,y) = \frac{5\sqrt{2}}{6\pi} \exp \left(-\frac{1}{2} \begin{bmatrix}\begin{pmatrix} x \\ y \end{pmatrix}  - \begin{pmatrix} 1 \\ 0 \end{pmatrix} \end{bmatrix}^{\intercal} * \begin{pmatrix} 11 & 5 \\ 5 & \frac{25}{9} \end{pmatrix} * \begin{bmatrix}\begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} 1 \\ 0 \end{pmatrix} \end{bmatrix} \right)$

```python
def p(x, y):
    return ((5 * np.sqrt(2))/ (6 * np.pi)) * np.exp(-0.5 * (11 * x ** 2 + 10 * x * y + (25/9) * y ** 2 - 22 * x - 10 * y + 11 ))

x = np.linspace(-3, 4, 200)
y = np.linspace(-5, 5, 200)

X, Y = np.meshgrid(x, y)
Z = p(X, Y)

fig = plt.figure(figsize=(10,8))
ax = plt.axes(projection='3d')

ax.plot_surface(X, Y, Z, cmap='rainbow')
ax.set_title('График функции плотности')
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('y', fontsize=12)
ax.set_zlabel('z', fontsize=12)
plt.show()
```
