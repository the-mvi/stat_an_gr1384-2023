### ${\xi}$ = $A$ ${\eta}$ + $b$.

### $\mathcal{p}_{\overrightarrow{\xi}}(x_1, \dots, x_n) = \frac{1}{(2\pi)^{n/2}\sqrt{\det \Sigma}} \exp(-\frac{1}{2}(\overrightarrow{x}-\overrightarrow{\mu})^T \Sigma^{-1}(\overrightarrow{x}-\overrightarrow{\mu}))$,

### где ${\mu}$ - математическое ожиданий случайного вектора ${\xi}$.

### Матрица ковариации:

### $\Sigma = \{\sigma_{i,j}\}_{i,j}^n$, где $\sigma_{i,j} = \mathop{cov}(\xi_i,\xi_j)$.

### 1. Вычислить вектор математических ожиданий и ковариационные характеристики данного случайного вектора. Построить график плотности распределения $p_{\xi, \eta}$.

### Нужно выражение в показателе экспоненты привести к следующему виду:
### $(x - \mu)^T\Sigma(x - \mu)$ или, что эквивалентно (очевидно после раскрытия данного матричного умножения) к квадратичной форме: $\sigma_1^2(x-\mu_1)^2 + 2\sigma_1\sigma_2(x-\mu_1)(x-\mu_2)+\sigma_2^2(x-\mu_2)^2$


### $3x^2 +4xy + 6y^2 -10x +12y +27=$

### $=(6y^2+12y+4yx) + 3x^2-10x+27=$

### $=6(y^2+2y(\frac{x+3}{3})) + 3x^2-10x+27=$

### $=6(y^2+2y(\frac{x+3}{3}) + (\frac{x+3}{3})^2) - 6(\frac{x+3}{3})^2 + 3x^2-10x+27 =$

### $=6(y^2+2y(\frac{x+3}{3}) + (\frac{x+3}{3})^2) + (\frac{7}{3}x^2-14x+21) =$

### $=6(y^2+2y(\frac{x+3}{3}) + (\frac{x+3}{3})^2) + \frac{7}{3}(x-3)^2 =$

### $=6(y + \frac{x}{3} + 1)^2 + \frac{7}{3}(x-3)^2 =$

### $=\frac{6}{9}((3y + 6)+(x-3))^2 + \frac{7}{3}(x-3)^2 =$

### $=\frac{6}{9}((3y +6)^2 + 2(3y+6)(x-3) + (x-3)^2) + \frac{7}{3}(x-3)^2 =$

### $=\frac{6}{9}(3y +6)^2 + 2 \cdot \frac{6}{9}(3y+6)(x-3) + \frac{6}{9}(x-3)^2 + \frac{7}{3}(x-3)^2 =$

### $= 3(x-3)^2 + 2 \cdot 2(y+2)(x-3) + 6(y + 2)^2$

### Выражение приведено к форме, теперь можно найти числовые характеристики:

### $\mu =  \begin{pmatrix} \mathbb{E}[\xi] \\ \mathbb{E}[\eta] \end{pmatrix} = \begin{pmatrix} 3 \\ -2 \end{pmatrix}$

### $\Sigma^{-1} = \begin{pmatrix} 3 & 2 \\ 2 & 6 \end{pmatrix} \Rightarrow \Sigma= \frac{1}{14}\begin{pmatrix} 6 & -2 \\ -2 & 3 \end{pmatrix}$

### Константа $C = \frac{1}{\sqrt{(2\pi)^2\cdot det(\Sigma)}} = \frac{1}{2\pi\sqrt{det(  \begin{pmatrix} \frac{3}{7} & -\frac{1}{7} \\ -\frac{1}{7} & \frac{3}{14} \end{pmatrix})}} = \frac{\sqrt{14}}{2\pi}$

### График плотности распределения:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
from mpl_toolkits.mplot3d import Axes3D

#Parameters to set
mu_x = 3
variance_x = 3

mu_y = -2
variance_y = 6

#Create grid and multivariate normal
x = np.linspace(-10,10,500)
y = np.linspace(-10,10,500)
X, Y = np.meshgrid(x,y)
pos = np.empty(X.shape + (2,))
pos[:, :, 0] = X; pos[:, :, 1] = Y
rv = multivariate_normal([mu_x, mu_y], [[variance_x, 2], [2, variance_y]])

#Make a 3D plot
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.plot_surface(X, Y, rv.pdf(pos),cmap='viridis',linewidth=0)
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
plt.show()
```

### 2. Найти аффинное ппреобразование, переводящее исходный случайный вектор в стандартный нормальный.

### Аффинное преобразование -  преобразование вида: $\Phi(x) = Ax + B$
### Найдем такие $A$ и $B$, чтобы они занулят мат. ожидания, а матрицу ковариации делали единичной. Иными словами, должно быть авыполнено:

### $A\mu+B=0$ и ${A\Sigma A} ^T = E$

### Возьмем одно из решений матричного уравнения: ${A\Sigma A} ^T = E$, а именно:

### $A = \begin{pmatrix} \sqrt{3} & \frac{2}{\sqrt{3}} \\ 0 & \frac{\sqrt{14}}{\sqrt{3}} \end{pmatrix}$

### Теперь найдем $B = -A\mu \Rightarrow$
### $B = \begin{pmatrix} -\sqrt{3} & -\frac{2}{\sqrt{3}} \\ 0 & -\frac{\sqrt{14}}{\sqrt{3}} \end{pmatrix}\begin{pmatrix}3 \\ -2\end{pmatrix} = \begin{pmatrix} -\frac{5}{\sqrt{3}} \\ 2 \cdot \sqrt{\frac{14}{3}}\end{pmatrix}$

### Тогда конечное аффинное преобразование, переводящее нормальный случайный вектор из задания в сьандартный нормальный вектор, выглядит так:

### $\Phi(x)  = \begin{pmatrix} \sqrt{3} & \frac{2}{\sqrt{3}} \\ 0 & \frac{\sqrt{14}}{\sqrt{3}} \end{pmatrix}x +  \begin{pmatrix} -\frac{5}{\sqrt{3}} \\ 2 \cdot \sqrt{\frac{14}{3}}\end{pmatrix}$

### Проверим, что данное преобразование переводит наш вектор в стандартный нормальный

### $\Phi\begin{pmatrix} \xi \\ \eta \end{pmatrix}  = \begin{pmatrix} \sqrt{3} & \frac{2}{\sqrt{3}} \\ 0 & \frac{\sqrt{14}}{\sqrt{3}} \end{pmatrix}\begin{pmatrix} \xi \\ \eta \end{pmatrix} +  \begin{pmatrix} -\frac{5}{\sqrt{3}} \\ 2 \cdot \sqrt{\frac{14}{3}}\end{pmatrix} = \begin{pmatrix} \sqrt{3}\xi +\frac{2}{\sqrt{3}}\eta -\frac{5}{\sqrt{3}} \\ \frac{\sqrt{14}}{\sqrt{3}}\cdot(\eta   + 2)\end{pmatrix}$

### Посчитаем числовые характеристики

### $\mu_{new} = \begin{pmatrix} \mathbb{E}[ \sqrt{3}\xi +\frac{2}{\sqrt{3}}\eta -\frac{5}{\sqrt{3}}] \\ \mathbb{E}[\frac{\sqrt{14}}{\sqrt{3}}\cdot(\eta   + 2)] \end{pmatrix}=$
### $\begin{pmatrix} 3 \cdot \sqrt{3} + -2 \cdot \frac{2}{\sqrt{3}} -\frac{5}{\sqrt{3}} \\ \frac{\sqrt{14}}{\sqrt{3}}\cdot(2-2) \end{pmatrix} = $
$\begin{pmatrix} 0 \\ 0 \end{pmatrix}$

### $\Sigma_{new}= {A\Sigma A} ^T = \begin{pmatrix} \sqrt{3} & \frac{2}{\sqrt{3}} \\ 0 & \frac{\sqrt{14}}{\sqrt{3}} \end{pmatrix}\begin{pmatrix} \frac{6}{14} & \frac{-2}{14} \\ \frac{-2}{14} & \frac{3}{14}\end{pmatrix}\begin{pmatrix} \sqrt{3} & 0 \\ \frac{2}{\sqrt{3}} & \frac{\sqrt{14}}{\sqrt{3}} \end{pmatrix} = \begin{pmatrix} 1 & 0 \\ 0 & 1 \end{pmatrix} = E$


### Действительно, новый вектор - стандартный нормальный.

### 3. Найти ортогональное преобразование, переводящее соответствующий цуентрированный случайный вектор в векотр с независимыми компонентами. Построить график плотности полученного распределения

### Чтобы найти такое преобразование необходимо вычислить собственные числа обратной матрицы ковариаций.

### $det(\Sigma^{-1} - \lambda E) = 0$
### $det(\begin{pmatrix} 3 - \lambda & 2 \\ 2 & 6 - \lambda \end{pmatrix}) = 0$
### $\lambda^2 - 9\lambda+14=0 \Rightarrow \lambda_1 = 2, \lambda_2 = 7$

### Собственные векторы и их отнормирование

### для $\lambda = 2$:
### $\begin{pmatrix} 1 & 2 \\ 2 & 4 \end{pmatrix}x_1 = 0 \Rightarrow x_1 = \frac{1}{\sqrt{5}}\begin{pmatrix}-2 \\ 1\end{pmatrix}$
### для $\lambda = 7$:
### $\begin{pmatrix} -4 & 2 \\ 2 & -1 \end{pmatrix}x_2 = 0 \Rightarrow x_2 = \frac{1}{\sqrt{5}}\begin{pmatrix}1 \\ 2\end{pmatrix}$

### Тогда итоговое ортогональное преобразование будет выглядеть так:
### $O = \frac{1}{\sqrt{5}}\begin{pmatrix} -2 & 1 \\ 1 & 2 \end{pmatrix}$

### Нужно убедиться, что преобразование действительно ортогональное:
### $\frac{1}{\sqrt{5}}\begin{pmatrix} -2 & 1 \\ 1 & 2 \end{pmatrix}={O}^T = {O}^{-1} = \frac{1}{\sqrt{5}}\begin{pmatrix} -2 & 1 \\ 1 & 2 \end{pmatrix}$

### $\begin{pmatrix}\xi' \\ \eta'\end{pmatrix} = O\begin{pmatrix}\xi \\ \eta\end{pmatrix} = \frac{1}{\sqrt{5}}\begin{pmatrix} -2 & 1 \\ 1 & 2 \end{pmatrix}\begin{pmatrix}\xi \\ \eta\end{pmatrix} = \frac{1}{\sqrt{5}}\begin{pmatrix} -2\xi +\eta \\ \xi + 2\eta \end{pmatrix}$

### Числовые характеристики нового распределения:
### $\Sigma' = {O\Sigma O}^T = \begin{pmatrix} \frac{1}{2} & 0 \\ 0 & \frac{1}{7} \end{pmatrix}$ , Отсюда видно, что случайные величины $\xi$ и $\eta$ независимы.

### $\mu' =  \begin{pmatrix} \mathbb{E}[\xi'] \\ \mathbb{E}[\eta'] \end{pmatrix} = \frac{1}{\sqrt5}\begin{pmatrix} -2\mathbb{E}[\xi] + \mathbb{E}[\eta] \\ \mathbb{E}[\xi] + 2\mathbb{E}[\eta] \end{pmatrix} = -\frac{1}{\sqrt5} \begin{pmatrix} 8\\ 1 \end{pmatrix}$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
from mpl_toolkits.mplot3d import Axes3D

#Parameters to set
mu_x = -(8/np.sqrt(5))
variance_x = 1/2

mu_y = -(1/np.sqrt(5))
variance_y = 1/7

#Create grid and multivariate normal
x = np.linspace(-10,10,500)
y = np.linspace(-10,10,500)
X, Y = np.meshgrid(x,y)
pos = np.empty(X.shape + (2,))
pos[:, :, 0] = X; pos[:, :, 1] = Y
rv = multivariate_normal([mu_x, mu_y], [[variance_x, 0], [0, variance_y]])

#Make a 3D plot
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.plot_surface(X, Y, rv.pdf(pos),cmap='viridis',linewidth=0)
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
plt.show()
```

### Числовые характеристики нового распределения:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
from mpl_toolkits.mplot3d import Axes3D

#Parameters to set
mu_x = -7
variance_x = 45/14

mu_y = 6
variance_y = 75/14

#Create grid and multivariate normal
x = np.linspace(-10,10,500)
y = np.linspace(-10,10,500)
X, Y = np.meshgrid(x,y)
pos = np.empty(X.shape + (2,))
pos[:, :, 0] = X; pos[:, :, 1] = Y
rv = multivariate_normal([mu_x, mu_y], [[variance_x, -55/14], [-55/14, variance_y]])

#Make a 3D plot
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.plot_surface(X, Y, rv.pdf(pos),cmap='viridis',linewidth=0)
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
plt.show()
```
