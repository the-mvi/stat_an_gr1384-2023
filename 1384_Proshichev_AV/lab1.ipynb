## Основные теоретические положения
**Определение**: \
Случайный вектор $ \vec{\xi}:\Omega \rightarrow  {\mathbb R}^n$, где $\vec{\xi} = (\xi_1 \dots \xi_n)^T$, называют нормальным (гауссовым), если существует вектор независимых стандартных случайных величин $ \vec{\eta} = (\eta_1 \dots \eta_m)^T$, матрица $A_{n \times m}$, и вещественный вектор $ \vec{B} = (b_1 \dots b_n)^T $, такие что: \
$ \vec{\xi}=A \vec{\eta} + \vec{B} $  
\
Если $ \vec{\xi} \sim \mathcal{N}(\vec{\mu}, \Sigma)$, где $ \vec{\mu} = \mathbb{E} \vec{\xi}$ и $\Sigma = \{cov(\xi_i, \xi_j)\}_{i = 1, j = 1}^n$, тогда плотность невырожденного многомерного нормального распределения будет иметь вид:\
$\rho_{\vec{\xi}}(x_1 \dots x_n) = \frac{1}{(2 \pi)^{n/_2} \sqrt{det \Sigma}} \exp(-\frac{1}{2} (\vec{x} - \vec{\mu})^T \Sigma^{-1} (\vec{x} - \vec{\mu}))$, где $ \vec{x} = (x_1 \dots x_n)^T $
\
\
**Свойства:** 
1) Линейная комбинация случайных переменных, пренадлежащих многомерному нормальному распределению, также является нормально распределённой. 
2) Условные и безусловные распределения, полученные из многомерного нормального распределения, также являются нормальными распределениями. 
3) Параметры многомерного нормального распределения польностью описывают его форму. 
4) Многомерное нормальное распределение устойчиво относительно линейных преобразования 
Если $\vec{\xi} \sim \mathcal{N}(\vec{\mu}, \Sigma)$, то $ A \vec{\xi} + \vec{B} \sim \mathcal{N}(A \vec{\mu} + \vec{B}, A \Sigma A^T) $

### **1 задание**. Вычислить вектор математических ожиданий и ковариационные характеристики данного случайного вектора. Построить график плотности распределения $ \rho_{\xi, \eta} $.
$\rho_{\xi, \eta}(x, y)=C \cdot \exp(-\frac{1}{2}(6x^2+4xy+9y^2+8x-14y+11)) $
\
\
Приведём квадратичную форму к каноническому виду. \
$ \sqsupset \varphi(x, y) = 6x^2+4xy+9y^2+8x-14y+11= (6x^2+4xy+8x) + 9y^2 - 14y + 11 = 2(3x^2 + 2x(y + 2)) + 9y^2 - 14y + 11 = 6(x^2 + 2x (\frac{y}{3} + \frac{2}{3})) + 9y^2 - 14y + 11 = 6(x^2 + 2x (\frac{y}{3} + \frac{2}{3}) + (\frac{y}{3} + \frac{2}{3})^2 - (\frac{y}{3} + \frac{2}{3})^2) + 9y^2 - 14y + 11 = 6(x + \frac{y}{3} + \frac{2}{3})^2 - 6(\frac{y}{3} + \frac{2}{3})^2 + 9y^2 - 14y + 11 = 6(x + \frac{y}{3} + \frac{2}{3})^2 - \frac{2y^2}{3} - \frac{8y}{3} - \frac{8}{3} + 9y^2 - 14y + 11 = 6(x + \frac{y}{3} + \frac{2}{3})^2 + \frac{25y^2}{3} - \frac{50y}{3} + \frac{25}{3} = 6(x + \frac{y}{3} + \frac{2}{3})^2 + \frac{25}{3}(y^2 - 2y + 1) = \mathbf{6(x + \frac{y}{3} + \frac{2}{3})^2 + \frac{25}{3}(y - 1)^2} = 6(x + \frac{1}{3}(y - 1 + 3))^2 + \frac{25}{3}(y - 1)^2 = 6((x + 1) + \frac{1}{3}(y - 1))^2 + \frac{25}{3}(y - 1)^2 = 6((x + 1)^2 + \frac{2}{3}(x + 1)(y - 1) + \frac{1}{9}(y - 1)^2) + \frac{25}{3}(y - 1)^2 = 6(x+1)^2 + 4(x + 1)(y - 1) + \frac{2}{3}(y - 1)^2 + \frac{25}{3}(y - 1)^2 = \mathbf{6(x+1)^2 + 4(x + 1)(y - 1) + 9(y - 1)^2}$
\
\
По каноническому виду мы можем найти мат. ожидание $\mathbb{E}\vec{\xi}$ и матрицу ковариаций $ \Sigma $. \
$\mathbb{E} \begin{pmatrix} \xi \\ \eta \end{pmatrix} = \begin{pmatrix} -1 \\ 1 \end{pmatrix} \quad \Sigma^{-1} = \begin{pmatrix} 6 & 2 \\ 2 & 9 \end{pmatrix} \quad det \Sigma^{-1} = 54 - 4 = 50 \implies \Sigma = \frac{1}{50} \begin{pmatrix} 9 & -2 \\ -2 & 6 \end{pmatrix}$ \
\
Найдём ковариационные характеристики данного случайного вектора
$ cov(\xi, \eta) = -\frac{1}{25} \quad \mathbb{D}_{\xi} = \frac{9}{50}  \quad \mathbb{D}_{\eta} = \frac{3}{25} \implies \rho(\xi, \eta)=\frac{cov({\xi},{ \eta})}{\sqrt{\mathbb{D}_{\xi} \mathbb{D}_{\eta}}} = \frac{-\frac{1}{25}}{\sqrt{\frac{9}{50} \frac{3}{25}}} = -\frac{25}{25} \frac{1}{\sqrt{\frac{27}{2}}} = -\frac{\sqrt{2}}{3 \sqrt{3}}$ \
\
Найдём коэффициент плотности распределения $C$. \
$C = \frac{1}{(2 \pi)^{\frac{n}{2}} \sqrt{det \Sigma}} = \frac{1}{2 \pi \sqrt{\frac{1}{50}}} = \frac{5 \sqrt{2}}{2 \pi}=\frac{5}{\pi \sqrt{2}}$ \
\
Тогда функция плотности выглядит следующим образом \
$\rho_{\xi, \eta}(x, y)= \frac{5}{\pi \sqrt{2}} \cdot \exp(-\frac{1}{2}(6x^2+4xy+9y^2+8x-14y+11)) $



```python
def p(x, y):
    return (5 / (np.pi * np.sqrt(2))) * np.exp(-0.5 * (6 * x ** 2 + 4 * x * y + 9 * y ** 2 + 8 * x - 14 * y + 11))


x = np.linspace(-3, 1, 200)
y = np.linspace(-3, 5, 200)

X, Y = np.meshgrid(x, y)
Z = p(X, Y)

fig = plt.figure(figsize=(10,8))
ax = plt.axes(projection='3d')

ax.plot_surface(X, Y, Z, cmap='OrRd')
ax.set_title('График функции плотности')
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('y', fontsize=12)
ax.set_zlabel('z', fontsize=12)
plt.show()
```
