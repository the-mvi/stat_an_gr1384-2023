#### 2) –ü–æ—Å—Ç—Ä–æ–∏—Ç—å –∏ –∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∏—Ä–æ–≤–∞—Ç—å –Ω–µ—Å–º–µ—â–µ–Ω–Ω—É—é –æ—Ü–µ–Ω–∫—É –¥–∏—Å–ø–µ—Ä—Å–∏–∏. –ù–∞ –±–∞–∑–µ –æ—à–∏–±–æ–∫ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—É —Å —à–∞–≥–æ–º h =0.79  –ü—Ä–æ–≤–µ—Ä–∏—Ç—å –≥–∏–ø–æ—Ç–µ–∑—É –Ω–æ—Ä–º–∞–ª—å–Ω–æ—Å—Ç–∏ –æ—à–∏–±–æ–∫ –Ω–∞ —É—Ä–æ–≤–Ω–µ –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ $\alpha$ = 0.02 –ø–æ $\chi^2$. –û—Ü–µ–Ω–∏—Ç—å —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –ø–æ–ª—É—á–µ–Ω–Ω–æ–π –æ—Ü–µ–Ω–∫–∏ –¥–æ –∫–ª–∞—Å—Å–∞ –Ω–æ—Ä–º–∞–ª—å–Ω—ã—Ö —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–π –ø–æ –ö–æ–ª–º–æ–≥–æ—Ä–æ–≤—É. –í–∏–∑—É–∞–ª—å–Ω–æ –æ—Ü–µ–Ω–∏—Ç—å –¥–∞–Ω–Ω—ã–π —Ñ–∞–∫—Ç.

–æ—Ü–µ–Ω–∫–∞ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ $s^2 =\frac{RSS(\hat{\beta})}{(n - r)}= \frac{1}{n-r} \cdot  (Y - X^T \widehat \beta)^T (Y - X^T \widehat \beta) $, –≥–¥–µ $r = rank(X) = 2$ <br>
$s^2 = 0.521$

```python
s = np.dot((Y_rez - np.dot(X_rez.T, beta)).T, (Y_rez - np.dot(X_rez.T, beta)))/(len(Y)-2)
print("–ù–µ—Å–º–µ—â–µ–Ω–Ω–∞—è –æ—Ü–µ–Ω–∫–∞ –¥–∏—Å–ø–µ—Ä—Å–∏–∏: ", s)
```

```python
Y_graphic = beta[0] + beta[1] * X_rez[1]
err = Y_rez - Y_graphic
h = 0.79
hist, bins, _ = plt.hist(err, bins=np.arange(min(err), max(err) + h, h), edgecolor='black', color='b')
plt.xlabel('–û—à–∏–±–∫–∏')
plt.ylabel('–ß–∞—Å—Ç–æ—Ç–∞')
plt.title('–û—Ü–µ–Ω–∫–∏ –æ—à–∏–±–æ–∫')
plt.show()
for i in range(len(hist)):
    print(f' –ò–Ω—Ç–µ—Ä–≤–∞–ª {i}: ({round(bins[i], 4)}, {round(bins[i+1], 4)}), –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —á–∏—Å–µ–ª –ø–æ–ø–∞–≤—à–∏—Ö –≤ –∏–Ω—Ç–µ—Ä–≤–∞–ª: {hist[i]}')
```

–ì–∏–ø–æ—Ç–µ–∑–∞ –Ω–∞ —É—Ä–æ–≤–Ω–µ –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ $\alpha = 0.02$ 

–û—Å–Ω–æ–≤–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞: $H_0 \sim \mathcal{N}(0, \sigma^2)$

–ü–µ—Ä–µ–æ–ø—Ä–µ–¥–µ–ª–∏–º –∏–Ω—Ç–µ—Ä–≤–∞–ª—ã, —á—Ç–æ–±—ã –≤ –∫–∞–∂–¥–æ–º –±—ã–ª–æ –Ω–µ –º–µ–Ω–µ–µ 5 –∑–Ω–∞—á–µ–Ω–∏–π:

$I_1 = (-\infty, -0.4618], \; v_1 = 10$

$I_2 = (-0.4618,  0.3282], \; v_2 = 27$

$I_3 = (0.3282, \infty), \; v_3 = 13$

–ü–æ—Å—á–∏—Ç–∞–µ–º —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫—É –∫—Ä–∏—Ç–µ—Ä–∏—è $\chi^2$:

–°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –∫—Ä–∏—Ç–µ—Ä–∏—è: $\chi^2 = \sum_{j=1}^{k}\frac{ (\nu_j - np_{jo})^2 }{ np_{j0} }$, –≥–¥–µ k - –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏–Ω—Ç–µ—Ä–≤–∞–ª–æ–≤

–ö–≤–∞–Ω—Ç–∏–ª—å $x_{\alpha}:$ $ \mathcal{K}_{1}(x_a) = 1 - \alpha_1 = 1 - 0.02 = 0.98$

```python
def minSum(a, g):
  array = [-0.4618,0.3282]
  res = [0] * (len(array) + 1)
  res[0] = sps.norm.cdf(array[0], a, g)
  res[1] = sps.norm.cdf(array[1], a, g) - sps.norm.cdf(array[0], a, g)
  res[2] = 1 - sps.norm.cdf(array[1], a, g)
  res = [round(elem, 5) for elem in res]
  v = [10, 27, 13]
  ans = [(v[i] - 50 * res[i]) ** 2 / (50 * res[i]) for i in range(len(v))]
  ans = [round(elem, 3) for elem in ans]
  return sum(ans)
print("X^2 =", minSum(0, s ** 0.5))
alpha = 0.02
x_alpha = stats.chi2(1).ppf(1 - alpha)
print(f'–ö–≤–∞–Ω—Ç–∏–ª—å x_alpha: {round(x_alpha, 4)}')
a_max = 1 - chi2.cdf(minSum(0, s ** 0.5), 1)
print("–ú–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π —É—Ä–æ–≤–µ–Ω—å –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ alpha, –ø—Ä–∏ –∫–æ—Ç–æ—Ä–æ–º –º–æ–∂–µ—Ç –±—ã—Ç—å –ø—Ä–∏–Ω—è—Ç–∞ –≥–∏–ø–æ—Ç–µ–∑–∞:", a_max)
```

$\chi^2 < x_\alpha \Rightarrow$ –ø—Ä–∏–Ω–∏–º–∞–µ–º $H_0$ 

–¢–µ–ø–µ—Ä—å –æ—Ü–µ–Ω–∏–º –≥–∏–ø–æ—Ç–µ–∑—É –ø—Ä–∏ –ø–æ–º–æ—â–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ö–æ–ª–º–∞–≥–æ—Ä–æ–≤–∞.

$ \sqrt{n} \cdot max|F_n(y) - F(y)| \sim K $,  $K$ - —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ö–æ–ª–º–æ–≥–æ—Ä–æ–≤–∞. 

```python
n = len(X)
def createFn(x, values):
    sum = 0
    for i in values:
        if i >= x:
            return sum / n
        sum += 1
    return sum / n
v = np.arange(np.min(err) - h, np.max(err) + h, h)
Fn = np.array([createFn(i, np.sort(err)) for i in v])
F = np.array([stats.norm.cdf(y, 0, np.sqrt(s)) for y in v])
Dn = np.max(np.abs(Fn - F)) * np.sqrt(n)
print("Dn: ", Dn)
x_alpha = stats.ksone.ppf(1 - alpha, n)
print("x_alpha: ", x_alpha)
```

$Dn > x_\alpha \Rightarrow$ –æ—Å–Ω–æ–≤–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ –æ—Ç–≤–µ—Ä–≥–∞–µ—Ç—Å—è

```python
plt.title('–§—É–Ω–∫—Ü–∏–∏ –≤—ã–±–æ—Ä–æ—á–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –æ—à–∏–±–∫–∏ –∏ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è')
plt.xlabel('x') 
plt.ylabel('–ó–Ω–∞—á–µ–Ω–∏—è —Ñ—É–Ω–∫—Ü–∏–π')
plt.plot(v, F, color='red')
plt.plot(v, Fn, color='blue')
plt.show()
plt.show()
```

#### 3. –í –ø—Ä–µ–¥–ø–æ–ª–æ–∂–µ–Ω–∏–∏ –Ω–æ—Ä–º–∞–ª—å–Ω–æ—Å—Ç–∏ –æ—à–∏–±–æ–∫ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–µ –∏–Ω—Ç–µ—Ä–≤–∞–ª—ã –¥–ª—è –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ $\beta_0$ –∏ $\beta_1$ —É—Ä–æ–≤–Ω—è –¥–æ–≤–µ—Ä–∏—è $1-\alpha$. –ü–æ—Å—Ç—Ä–æ–∏—Ç—å –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π —ç–ª–ª–∏–ø—Å —É—Ä–æ–≤–Ω—è –¥–æ–≤–µ—Ä–∏—è $1-\alpha$ –¥–ª—è $(\beta_0, \beta_1)$ (–≤—ã—á–∏—Å–ª–∏—Ç—å –µ–≥–æ –ø–æ–ª—É–æ—Å–∏).

–î–ù–û —Ñ—É–Ω–∫—Ü–∏—è: $\psi = C^T \beta $

–≠–ª–µ–º–µ–Ω—Ç –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –æ—Ü–µ–Ω–∫–∏: $ b_\psi = C^T (XX^T)^{-1}C$, –≥–¥–µ  $C$ - –º–∞—Ç—Ä–∏—Ü–∞ –¥–ª—è –æ–ø–µ—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤, –∫–æ—Ç–æ—Ä—ã–µ –±—É–¥—É—Ç –æ—Ü–µ–Ω–µ–Ω—ã.

–¢–æ–≥–¥–∞ –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞: $ \cfrac{\widehat \psi - \psi}{s\sqrt{b}} \sim S_{n-r} $

–î–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª —É—Ä–æ–≤–Ω—è –¥–æ–≤–µ—Ä–∏—è $1 - \alpha$:

$ P_\theta(-x_\alpha \le \cfrac{\widehat \psi - \psi}{s\sqrt{b}} \le x_\alpha) = 1 -\alpha $, <br>
–≥–¥–µ –∫–≤–∞–Ω—Ç–∏–ª—å $ x_\alpha : S_{n-r}(x_\alpha) = 1 -\frac{\alpha}{2} $

$b_{\psi_0} = C_0^T (X X^T)^{-1} C_0 = 0.0625$


$b_{\psi_1} = C_1^T (X X^T)^{-1} C_1 = 0.0162$

–ù–∞–π–¥–µ–º $x_{\alpha} = S^{-1}_{n - 2}(1 - \frac{\alpha}{2}) = 2.407$

```python
x_alpha = stats.t.ppf(1-alpha/2, n-2)
print("x_alpha: ", round(x_alpha, 3))
b_0 = np.dot(np.dot(np.array([1, 0]).T, np.linalg.inv(XXT)),np.array([1, 0]))
print("b0:", round(b_0, 3))
b_1 = np.dot(np.dot(np.array([0, 1]).T,np.linalg.inv(XXT)), np.array([0, 1]))
print("b1: ", round(b_1, 3))
print(f'–î–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª [{round(beta[0] - x_alpha * np.sqrt(s) * np.sqrt(b_0), 3)}, {round(beta[0] + x_alpha * np.sqrt(s) * np.sqrt(b_0), 3)}] –¥–ª—è beta_0')
print(f'–î–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª [{round(beta[1] - x_alpha * np.sqrt(s) * np.sqrt(b_1), 3)}, {round(beta[1] + x_alpha * np.sqrt(s) * np.sqrt(b_1), 3)}] –¥–ª—è beta_1')
```

–î–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π —ç–ª–ª–∏–ø—Å–æ–∏–¥ —É—Ä–æ–≤–Ω—è –¥–æ–≤–µ—Ä–∏—è 1 - $\alpha$ :

$ A_\alpha = \{x: (x - \widehat \psi)^T (C^T (XX^T)^{-1}C)^{-1}(x - \widehat \psi) \le qs^2x_\alpha\}$, –≥–¥–µ $ x_\alpha = F_{q, n-r}(x_\alpha) = 1 - \alpha$

$C = \begin{pmatrix}
    1 & 0\\
    0 & 1
\end{pmatrix} $

–í –∫–∞—á–µ—Å—Ç–≤–µ –≤–µ–∫—Ç–æ—Ä–∞ –≤–æ–∑—å–º–µ–º $ \vec{x}= P(\hat{\psi} - \psi) $, –≥–¥–µ P - –º–∞—Ç—Ä–∏—Ü–∞ —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã—Ö –≤–µ–∫—Ç–æ—Ä–æ–≤ 

–£—Ä–∞–≤–Ω–µ–Ω–∏–µ —ç–ª–∏–ø—Å–æ–π–¥–∞

$(\widehat \psi -  \psi)^T (C^T (XX^T)^{-1}C)^{-1}(\widehat \psi -  \psi) = \lambda_1 x^2 + \lambda_2 y^2 \le qs^2x_\alpha$, –≥–¥–µ $\lambda_1$ –∏ $\lambda_2$ - —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã–µ —á–∏—Å–ª–∞ –º–∞—Ç—Ä–∏—Ü—ã $XX^T$

$\cfrac{\lambda_1}{qs^2x_\alpha}x^2 + \cfrac{\lambda_2}{qs^2x_\alpha}y^2 \le 1$

–ë–æ–ª—å—à–∞—è –∏ –º–∞–ª–∞—è –ø–æ–ª—É–æ—Å–∏ –º–æ–≥—É—Ç –±—ã—Ç—å –≤—ã—á–∏—Å–ª–µ–Ω—ã –∫–∞–∫ $ \frac{q s^2 x_{\alpha}}{\lambda_1} $ –∏ $ \frac{q s^2 x_{\alpha}}{\lambda_2} $ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ.

$ \vec{x}= P(\hat{\psi} - \psi) = \begin{pmatrix} x\\ y \end{pmatrix} = \begin{pmatrix} -0.91153022 && -0.4112331 \\ 0.4112331 && -0.91153022 \end{pmatrix} \cdot \begin{pmatrix} 8.27690191 - \beta_0 \\ -0.23512464 - \beta_1 \end{pmatrix}$ 

```python
evalues, evectors = np.linalg.eig(XXT)
x_alpha = stats.f.ppf(1-alpha, 2, n-2)
print("x_alpha: ", x_alpha)

a = np.sqrt((2 * s * x_alpha) / evalues[0])
b =  np.sqrt((2 * s * x_alpha) / evalues[1])

print("–ë–æ–ª—å—à–∞—è –ø–æ–ª—É–æ—Å—å: ", round(a, 3))
print("–ú–∞–ª–∞—è –ø–æ–ª—É–æ—Å—å: ", round(b, 3))

print("beta: ", beta)

def ellipse_equation(beta0, beta1):
    return (( evectors[0][0] * (beta[0] - beta0) - evectors[0][1] * (beta[1] - beta1))**2 / a**2) + ((evectors[1][0] * (beta[0] - beta0) + evectors[1][1] * (beta[1] - beta1))**2 / b**2)


print("–°–æ–±—Å—Ç–≤–µ–Ω–Ω—ã–µ –≤–µ–∫—Ç–æ—Ä–∞: ", evectors)
print("–°–æ–±—Å—Ç–≤–µ–Ω–Ω—ã–µ —á–∏—Å–ª–∞: ", evalues)
x = np.linspace(5, 10, 200)
y = np.linspace(-1, 1, 200)
X_rez, Y_rez = np.meshgrid(x, y)
plt.figure(figsize=(6, 6))
plt.contour(X_rez, Y_rez,ellipse_equation(X_rez, Y_rez), levels=[1], colors='blue', linewidths=2)

plt.title("–≠–ª–ª–∏–ø—Å")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()
```

#### 4) –°—Ñ–æ—Ä–º—É–ª–∏—Ä–æ–≤–∞—Ç—å –≥–∏–ø–æ—Ç–µ–∑—É –Ω–µ–∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π Y –æ—Ç –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π X. –ü—Ä–æ–≤–µ—Å—Ç–∏ –ø—Ä–æ–≤–µ—Ä–∫—É –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏.
–î–ù–û - —Ñ—É–Ω–∫—Ü–∏—è: 
$\psi = C^T \beta $


–û—Å–Ω–æ–≤–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞:
$ H_0 : \beta_1 = 0$

Y –Ω–µ –∑–∞–≤–∏—Å–∏—Ç –æ—Ç X, –µ—Å–ª–∏ $ y_i \approx \beta_0 + \beta_1 x_i = \beta_0 $\
–¢–æ –µ—Å—Ç—å $ \beta_1 = 0 $

$ C = \left( \begin{array}{c} 0 \\ 1 \end{array} \right) \implies \psi = \left( \begin{array}{c} 0 & 1 \end{array} \right) \left( \begin{array}{c} \beta_0 \\ \beta_1 \end{array} \right) = \beta_1$

$\mathcal{F} = \cfrac{\widehat \psi^T ((C^T (XX^T)^{-1}C)^{-1}) \widehat \psi}{qs^2} \sim F_{q, n-r}$

$\mathcal{F} = \cfrac{{\widehat \beta_1} ^2}{bqs^2} $

$x_\alpha: F_{q, n-r}(x_\alpha) = 1 - \alpha $




```python
b = np.dot(np.dot( np.array([0, 1]).T, np.linalg.inv(XXT)),  np.array([0, 1]))
print("b: ", round(b, 3))
F = beta[1] ** 2 / (b * 1 * s)
print("F: ", round(F, 3))
print("alpha: ", alpha)
x_alpha = stats.f(1, n - 2).ppf(1 - alpha)
print("x_alpha: ", round(x_alpha, 4))
a_max = 1 - stats.f.cdf(F, 1, n - 2)
print("–ú–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π —É—Ä–æ–≤–µ–Ω—å –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ alpha, –ø—Ä–∏ –∫–æ—Ç–æ—Ä–æ–º –º–æ–∂–µ—Ç –±—ã—Ç—å –ø—Ä–∏–Ω—è—Ç–∞ –≥–∏–ø–æ—Ç–µ–∑–∞:", a_max)
```

 $\mathcal{F} > x_\alpha$, –æ—Ç–≤–µ—Ä–≥–∞–µ–º $H_0$ 

#### 5) –°—Ñ–æ—Ä–º—É–ª–∏—Ä–æ–≤–∞—Ç—å –º–æ–¥–µ–ª—å, –≤–∫–ª—é—á–∞—é—â—É—é –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–π —á–ª–µ–Ω $X^2$. –ü–æ—Å—Ç—Ä–æ–∏—Ç—å –ú–ù–ö –æ—Ü–µ–Ω–∫–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ $\beta_0, \beta_1, \beta_2$ –≤ –¥–∞–Ω–Ω–æ–π –º–æ–¥–µ–ª–∏. –ò–∑–æ–±—Ä–∞–∑–∏—Ç—å –ø–æ–ª—É—á–µ–Ω–Ω—É—é —Ä–µ–≥—Ä–µ—Å—Å–∏–æ–Ω–Ω—É—é –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å.
 1)$\mathbb{E}_\theta e = 0$ 

 2)$Var_\theta e = \sigma^2 E$

1-2 —Å–ª–µ–¥—É—é—Ç –∏–∑ —Ä–µ–≥—Ä–µ—Å—Å–∏–æ–Ω–Ω–æ–π –º–æ–¥–µ–ª–∏ 

–°–∏—Å—Ç–µ–º–∞ –Ω–æ—Ä–º–∞–ª—å–Ω—ã—Ö —É—Ä–∞–≤–Ω–µ–Ω–∏–π
$ X X^T \beta = X Y $

–ú–ù–ö –æ—Ü–µ–Ω–∫–∞:

$ \hat{\beta} = (X X^T)^{-1} XY $

$Y = \beta_0 + \beta_1 X + \beta_1X^2 + \mathbb \varepsilon$

$\beta = (\beta_0, \beta_1, \beta_2) $


```python
X_rez = np.array([[1 for i in range(50)], X, [i**2 for i in X]])
Y_rez = np.array(Y)
XXT = np.dot(X_rez, np.transpose(X_rez))
print("XXT: ", XXT)
XXT_inv = np.linalg.inv(XXT)
print("XXT_inv: ", XXT_inv)
beta = np.dot(np.dot(XXT_inv, X_rez), Y_rez)

print("b0:", round(beta[0], 3), "b1:", round(beta[1], 3), "b2:", round(beta[2], 3))
```

#### 6) –ü–æ—Å—Ç—Ä–æ–∏—Ç—å –Ω–µ—Å–º–µ—â–µ–Ω–Ω—É—é –æ—Ü–µ–Ω–∫—É –¥–∏—Å–ø–µ—Ä—Å–∏–∏. –ü—Ä–æ–≤–µ—Å—Ç–∏ –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏–µ –Ω–æ—Ä–º–∞–ª—å–Ω–æ—Å—Ç–∏ –æ—à–∏–±–æ–∫ –∫–∞–∫ –≤ –ø—É–Ω–∫—Ç–µ 2.

–æ—Ü–µ–Ω–∫–∞ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ $s^2 =\frac{RSS(\hat{\beta})}{(n - r)}= \frac{1}{n-r} \cdot  (Y - X^T \widehat \beta)^T (Y - X^T \widehat \beta) $, –≥–¥–µ $r = rank(X) = 3$ 

$s^2 = 0.55$

```python
s = np.dot((Y_rez - np.dot(X_rez.T, beta)).T, (Y_rez - np.dot(X_rez.T, beta)))/(len(Y)-3)
print("–ù–µ—Å–º–µ—â–µ–Ω–Ω–∞—è –æ—Ü–µ–Ω–∫–∞ –¥–∏—Å–ø–µ—Ä—Å–∏–∏: ", s)
```

```python
Y_graphic = beta[0] + beta[1] * X_rez[1] + beta[2]*X_rez[1]**2
err = Y_rez - Y_graphic
h = 0.79
hist, bins, _ = plt.hist(err, bins=np.arange(min(err), max(err) + h, h), edgecolor='black', color='b')
plt.xlabel('–û—à–∏–±–∫–∏')
plt.ylabel('–ß–∞—Å—Ç–æ—Ç–∞')
plt.title('–û—Ü–µ–Ω–∫–∏ –æ—à–∏–±–æ–∫')
plt.show()
for i in range(len(hist)):
    print(f' –ò–Ω—Ç–µ—Ä–≤–∞–ª {i}: ({round(bins[i], 4)}, {round(bins[i+1], 4)}), –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —á–∏—Å–µ–ª –ø–æ–ø–∞–≤—à–∏—Ö –≤ –∏–Ω—Ç–µ—Ä–≤–∞–ª: {hist[i]}')
```

–ì–∏–ø–æ—Ç–µ–∑–∞ –Ω–∞ —É—Ä–æ–≤–Ω–µ –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ $\alpha = 0.02$ 

–û—Å–Ω–æ–≤–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞: $H_0 \sim \mathcal{N}(0, \sigma^2)$

–ü–µ—Ä–µ–æ–ø—Ä–µ–¥–µ–ª–∏–º –∏–Ω—Ç–µ—Ä–≤–∞–ª—ã, —á—Ç–æ–±—ã –≤ –∫–∞–∂–¥–æ–º –±—ã–ª–æ –Ω–µ –º–µ–Ω–µ–µ 5 –∑–Ω–∞—á–µ–Ω–∏–π:

$I_1 = (-\infty, -0.5224], \; v_1 = 10$

$I_2 = (-0.5224,  0.2676], \; v_2 = 24$

$I_3 = (0.2676, \infty), \; v_3 = 16$

–ü–æ—Å—á–∏—Ç–∞–µ–º —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫—É –∫—Ä–∏—Ç–µ—Ä–∏—è $\chi^2$:

–°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –∫—Ä–∏—Ç–µ—Ä–∏—è: $\chi^2 = \sum_{j=1}^{k}\frac{ (\nu_j - np_{jo})^2 }{ np_{j0} }$, –≥–¥–µ k - –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏–Ω—Ç–µ—Ä–≤–∞–ª–æ–≤

–ö–≤–∞–Ω—Ç–∏–ª—å $x_{\alpha}:$ $ \mathcal{K}_{0}(x_a) = 1 - \alpha_1 = 1 - 0.02 = 0.98$

```python
def minSum(a, g):
  array = [-0.5224,0.2676]
  res = [0] * (len(array) + 1)
  res[0] = sps.norm.cdf(array[0], a, g)
  res[1] = sps.norm.cdf(array[1], a, g) - sps.norm.cdf(array[0], a, g)
  res[2] = 1 - sps.norm.cdf(array[1], a, g)
  res = [round(elem, 5) for elem in res]
  v = [10, 24, 16]
  ans = [(v[i] - 50 * res[i]) ** 2 / (50 * res[i]) for i in range(len(v))]
  ans = [round(elem, 3) for elem in ans]
  return sum(ans)
print("X^2 =", minSum(0, s ** 0.5))
alpha = 0.02
x_alpha = stats.chi2(1).ppf(1 - alpha)
print("–ö–≤–∞–Ω—Ç–∏–ª—å x_alpha: ", round(x_alpha, 3))
a_max = 1 - chi2.cdf(minSum(0, s ** 0.5), 1)
print("–ú–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π —É—Ä–æ–≤–µ–Ω—å –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ alpha, –ø—Ä–∏ –∫–æ—Ç–æ—Ä–æ–º –º–æ–∂–µ—Ç –±—ã—Ç—å –ø—Ä–∏–Ω—è—Ç–∞ –≥–∏–ø–æ—Ç–µ–∑–∞:", a_max)
```

$\chi^2 < x_\alpha \Rightarrow$ –ø—Ä–∏–Ω–∏–º–∞–µ–º $H_0$ 

–¢–µ–ø–µ—Ä—å –æ—Ü–µ–Ω–∏–º –≥–∏–ø–æ—Ç–µ–∑—É –ø—Ä–∏ –ø–æ–º–æ—â–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ö–æ–ª–º–∞–≥–æ—Ä–æ–≤–∞.

$ \sqrt{n} \cdot max|F_n(y) - F(y)| \sim K $,  $K$ - —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ö–æ–ª–º–æ–≥–æ—Ä–æ–≤–∞. 

```python
n = len(X)
def createFn(x, values):
    sum = 0
    for i in values:
        if i >= x:
            return sum / n
        sum += 1
    return sum / n
v = np.arange(np.min(err) - h, np.max(err) + h, h)
Fn = np.array([createFn(i, np.sort(err)) for i in v])
F = np.array([stats.norm.cdf(y, 0, np.sqrt(s)) for y in v])
Dn = np.max(np.abs(Fn - F)) * np.sqrt(n)
print("Dn: ", Dn)
x_alpha = stats.ksone.ppf(1 - alpha, n)
print("x_alpha: ", x_alpha)
```

$Dn > x_\alpha \Rightarrow$ –æ—Å–Ω–æ–≤–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ –æ—Ç–≤–µ—Ä–≥–∞–µ—Ç—Å—è

```python
plt.title('–§—É–Ω–∫—Ü–∏–∏ –≤—ã–±–æ—Ä–æ—á–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –æ—à–∏–±–∫–∏ –∏ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è')
plt.xlabel('x') 
plt.ylabel('–ó–Ω–∞—á–µ–Ω–∏—è —Ñ—É–Ω–∫—Ü–∏–π')
plt.plot(v, F, color='red')
plt.plot(v, Fn, color='blue')
plt.show()
plt.show()
```

#### 7) –í –ø—Ä–µ–¥–ø–æ–ª–æ–∂–µ–Ω–∏–∏ –Ω–æ—Ä–º–∞–ª—å–Ω–æ—Å—Ç–∏ –æ—à–∏–±–æ–∫ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–µ –∏–Ω—Ç–µ—Ä–≤–∞–ª—ã –¥–ª—è –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ $\beta_0$, $\beta_1$, $\beta_2$  —É—Ä–æ–≤–Ω—è –¥–æ–≤–µ—Ä–∏—è 1 ‚àí $\alpha$. –ù–∞–ø–∏—Å–∞—Ç—å —É—Ä–∞–≤–Ω–µ–Ω–∏–µ –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω–æ–≥–æ —ç–ª–ª–∏–ø—Å–æ–∏–¥–∞ —É—Ä–æ–≤–Ω—è –¥–æ–≤–µ—Ä–∏—è 1 ‚àí $\alpha$.
–î–ù–û —Ñ—É–Ω–∫—Ü–∏—è: $\psi = C^T \beta $

–≠–ª–µ–º–µ–Ω—Ç –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –æ—Ü–µ–Ω–∫–∏: $ b_\psi = C^T (XX^T)^{-1}C$, –≥–¥–µ  $C$ - –º–∞—Ç—Ä–∏—Ü–∞ –¥–ª—è –æ–ø–µ—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤, –∫–æ—Ç–æ—Ä—ã–µ –±—É–¥—É—Ç –æ—Ü–µ–Ω–µ–Ω—ã.

–¢–æ–≥–¥–∞ –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞: $ \cfrac{\widehat \psi - \psi}{s\sqrt{b}} \sim S_{n-r} $

–î–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª —É—Ä–æ–≤–Ω—è –¥–æ–≤–µ—Ä–∏—è $1 - \alpha$:

$ P_\theta(-x_\alpha \le \cfrac{\widehat \psi - \psi}{s\sqrt{b}} \le x_\alpha) = 1 -\alpha $, <br>
–≥–¥–µ –∫–≤–∞–Ω—Ç–∏–ª—å $ x_\alpha : S_{n-r}(x_\alpha) = 1 -\frac{\alpha}{2} $

$b_{\psi_0} = C_0^T (X X^T)^{-1} C_0 = 0.086$


$b_{\psi_1} = C_1^T (X X^T)^{-1} C_1 = 0.203$

$b_{\psi_2} = C_2^T (X X^T)^{-1} C_2 = 0.02$

–ù–∞–π–¥–µ–º $x_{\alpha} = S^{-1}_{n - 3}(1 - \frac{\alpha}{2}) = 2.408$

```python
x_alpha = stats.t.ppf(1-alpha/2, n-3)
print("x_alpha: ", round(x_alpha, 3))
b_0 = np.dot(np.dot(np.array([1, 0, 0]).T, np.linalg.inv(XXT)),np.array([1, 0, 0]))
print("b0:", round(b_0, 3))
b_1 = np.dot(np.dot(np.array([0, 1, 0]).T,np.linalg.inv(XXT)), np.array([0, 1, 0]))
print("b1: ", round(b_1, 3))
b_2 = np.dot(np.dot(np.array([0, 0, 1]).T,np.linalg.inv(XXT)), np.array([0, 0, 1]))
print("b2: ", round(b_2, 3))
print(f'–î–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª [{round(beta[0] - x_alpha * np.sqrt(s) * np.sqrt(b_0), 3)}, {round(beta[0] + x_alpha * np.sqrt(s) * np.sqrt(b_0), 3)}] –¥–ª—è beta_0')
print(f'–î–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª [{round(beta[1] - x_alpha * np.sqrt(s) * np.sqrt(b_1), 3)}, {round(beta[1] + x_alpha * np.sqrt(s) * np.sqrt(b_1), 3)}] –¥–ª—è beta_1')
print(f'–î–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª [{round(beta[2] - x_alpha * np.sqrt(s) * np.sqrt(b_2), 3)}, {round(beta[2] + x_alpha * np.sqrt(s) * np.sqrt(b_2), 3)}] –¥–ª—è beta_2')
```

–î–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π —ç–ª–ª–∏–ø—Å–æ–∏–¥ —É—Ä–æ–≤–Ω—è –¥–æ–≤–µ—Ä–∏—è 1 - $\alpha$ :

$ A_\alpha = \{x: (x - \widehat \psi)^T (C^T (XX^T)^{-1}C)^{-1}(x - \widehat \psi) \le qs^2x_\alpha\}$, –≥–¥–µ $ x_\alpha = F_{q, n-r}(x_\alpha) = 1 - \alpha$

$C = \begin{pmatrix}
    1 & 0 & 0\\
    0 & 1 & 0\\
    0 & 0 & 1
\end{pmatrix} $

–í –∫–∞—á–µ—Å—Ç–≤–µ –≤–µ–∫—Ç–æ—Ä–∞ –≤–æ–∑—å–º–µ–º $ \vec{x}= P(\hat{\psi} - \psi) $, –≥–¥–µ P - –º–∞—Ç—Ä–∏—Ü–∞ —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã—Ö –≤–µ–∫—Ç–æ—Ä–æ–≤ 

–£—Ä–∞–≤–Ω–µ–Ω–∏–µ —ç–ª–∏–ø—Å–æ–π–¥–∞

$(\widehat \psi -  \psi)^T (C^T (XX^T)^{-1}C)^{-1}(\widehat \psi -  \psi) = \lambda_1 x^2 + \lambda_2 y^2 + \lambda_3 z^2 \le qs^2x_\alpha$, –≥–¥–µ $\lambda_1$ –∏ $\lambda_2$ $\lambda_3$- —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã–µ —á–∏—Å–ª–∞ –º–∞—Ç—Ä–∏—Ü—ã $XX^T$

$\cfrac{\lambda_1}{qs^2x_\alpha}x^2 + \cfrac{\lambda_2}{qs^2x_\alpha}y^2 + \cfrac{\lambda_3}{qs^2x_\alpha}z^2\le 1$

$1185.768 x^2 +  19.682 y^2 +  2.778 z^2\le 1$

–ë–æ–ª—å—à–∞—è –∏ –º–∞–ª–∞—è –ø–æ–ª—É–æ—Å–∏ –º–æ–≥—É—Ç –±—ã—Ç—å –≤—ã—á–∏—Å–ª–µ–Ω—ã –∫–∞–∫ $ \frac{q s^2 x_{\alpha}}{\lambda_1} $ –∏ $ \frac{q s^2 x_{\alpha}}{\lambda_2} $ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ.

$ \vec{x}= P(\hat{\psi} - \psi) = \begin{pmatrix} x \\ y \\ z\end{pmatrix} =\begin{pmatrix} 
-0.13523689 & -0.87904031 & 0.45716422\\ 
-0.34313982 & -0.39129755 &-0.85389771\\
-0.92949774 & 0.27234972 & 0.24871587
\end{pmatrix} 
\cdot \begin{pmatrix} 
8.20951273 - \beta_0 \\ 
-0.04466727 - \beta_1 \\
-0.06247455 - \beta_2
\end{pmatrix}$ 


```python
evalues, evectors = np.linalg.eig(XXT)
x_alpha = stats.f.ppf(1-alpha, 2, n-3)
print("x_alpha: ", x_alpha)

a = np.sqrt((2 * s * x_alpha) / evalues[0])
b =  np.sqrt((2 * s * x_alpha) / evalues[1])

print("–ë–æ–ª—å—à–∞—è –ø–æ–ª—É–æ—Å—å: ", round(a, 3))
print("–ú–∞–ª–∞—è –ø–æ–ª—É–æ—Å—å: ", round(b, 3))

print("beta: ", beta)
print("–°–æ–±—Å—Ç–≤–µ–Ω–Ω—ã–µ –≤–µ–∫—Ç–æ—Ä–∞:\n ", evectors)
print("–°–æ–±—Å—Ç–≤–µ–Ω–Ω—ã–µ —á–∏—Å–ª–∞:\n ", evalues)

array = []
for v in evalues:
    array.append(v / 3 * s * x_alpha)
print(round(array[0] , 3),"x^2 + ", round(array[1] , 3),"y^2 + ", round(array[2] , 3),"z^2")
```

#### 8) –°—Ñ–æ—Ä–º—É–ª–∏—Ä–æ–≤–∞—Ç—å –≥–∏–ø–æ—Ç–µ–∑—É –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–Ω–Ω–æ–π –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π Y –æ—Ç –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π X –∏ –ø—Ä–æ–≤–µ—Ä–∏—Ç—å –µ—ë –∑–Ω–∞—á–∏–º–æ—Å—Ç—å –Ω–∞ —É—Ä–æ–≤–Ω–µ $\alpha$.h.e) 
–î–ù–û - —Ñ—É–Ω–∫—Ü–∏—è: 
$\psi = C^T \beta $


–û—Å–Ω–æ–≤–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞:
$ H_0 : \beta_1 = 0$

Y –Ω–µ –∑–∞–≤–∏—Å–∏—Ç –æ—Ç X, –µ—Å–ª–∏ $ y_i \approx \beta_0 + \beta_1 x_i = \beta_0 $\
–¢–æ –µ—Å—Ç—å $ \beta_1 = 0 $

$ C = \left( \begin{array}{c} 0 \\ 0 \\ 1\end{array} \right) \implies \psi = \left( \begin{array}{c} 0 & 0 & 1 \end{array} \right) \left( \begin{array}{c} \beta_0 \\ \beta_1 \end{array} \right) = \beta_1$

$\mathcal{F} = \cfrac{\widehat \psi^T ((C^T (XX^T)^{-1}C)^{-1}) \widehat \psi}{qs^2} \sim F_{q, n-r}$

$\mathcal{F} = \cfrac{{\widehat \beta_2} ^2}{bqs^2} $

$x_\alpha: F_{q, n-r}(x_\alpha) = 1 - \alpha $


```python
b = np.dot(np.dot( np.array([0, 0, 1]).T, np.linalg.inv(XXT)),  np.array([0, 0, 1]))
print("b: ", round(b, 3))
F = beta[1] ** 2 / (b * 1 * s)
print("F: ", round(F, 3))
print("alpha: ", alpha)
x_alpha = stats.f(1, n - 2).ppf(1 - alpha)
print("x_alpha: ", round(x_alpha, 4))
a_max = 1 - stats.f.cdf(F, 1, n - 2)
print("–ú–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π —É—Ä–æ–≤–µ–Ω—å –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ alpha, –ø—Ä–∏ –∫–æ—Ç–æ—Ä–æ–º –º–æ–∂–µ—Ç –±—ã—Ç—å –ø—Ä–∏–Ω—è—Ç–∞ –≥–∏–ø–æ—Ç–µ–∑–∞:", a_max)
```
