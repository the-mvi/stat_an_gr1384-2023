## Основные теоретические положения
Случайный вектор $ \overline\xi = (\xi_1,\, ...,\, \xi_n)^{\text{T}}\, :\, \Omega \rightarrow \mathbb{R}^n $ - называют нормальным (Гауссовским), если существует вектор независимых стандартных нормальных случайных величин $\overline{\eta} = (\eta_1,\, ...,\, \eta_m)^{\text{T}},$ матрица $A$ размерности $ m \times n $ и вещественный вектор $ B = (b_1,\, ...\, b_n)^\text{T}, $ такие что:
$$
    \overline{\xi} = A\overline{\eta} + B.
$$
Причем $ \overline\xi \sim \mathcal{N}(\mu, \Sigma), $ где матожидание $ \mu = \mathbb{E}\overline\xi = (E\xi_1,\, ...,\, E\xi_n)^{\text{T}}, $ ковариационная матрица $ \Sigma = \{\sigma_{ij}\}_{i,j = 1}^{n},\, \sigma_{ij} = \mathop{\mathrm{cov}}(\xi_i,\, \xi_j), $ а плотность распределения имеет вид:
$$
    \begin{equation}
        \rho_{\overline{\xi}} = \frac{1}{(2\pi)^{\frac{n}{2}}\sqrt{\det{\Sigma}}}\exp\left ( -\frac{1}{2}\left(\overline{x}-\mu\right)^\text{T}\Sigma^{-1}\left(\overline{x}-\mu\right)\right ), \\
        \overline{x} = (x_1,\, ...,\, x_n)^\text{T}.
    \end{equation}
$$
- **Математическим ожиданием** случайной величины $\xi$ называют **число** $\mathbb{E}\xi$ определяемое равенством $$\mathbb{E}\xi = \int_{\omega\in\Omega} \xi(\omega) \, dP(\omega) = \int_{-\infty}^{\infty} x \, dF_{\xi}(x)$$ где $F\substack(\xi)$ - функция распределения величины $\xi$. Если интеграл расходится, то говорят, что у случайной величины отсутствует математическое ожидание. Для дискретной случайной величины $\mathbb{E}\xi=\sum_{i;p_{i}>0}a_{i}p_{i}, p_{i}=P(\xi = a_{i}), \sum_{i}p_{i} = 1$ Для абсолютно непрерывной случайной величины $\mathbb{E}\xi = \int_{\mathbb{R}} xp_{\xi}(x) \, dx$.
- **Дисперсией** случайной величины $\xi$ называется центральный момент второго порядка $\mathbb{E}(\xi-\mathbb{E}\xi)^2$, если этот момент существует.
<br>Стандартное обозначение для дисперсии - $\mathbb{D}\xi$.
<br>Из определения математического ожидания следует, что $$\mathbb{D}\xi = \int_{\Omega} (\xi(\omega)-\mathbb{E}\xi)^2 \, dP(\omega)$$
Таким образом, дисперсия - это среднее значение квадрата отклонения случайной величины от её математического ожидания.
- **Ковариационная характеристика** используются для описания связи или зависимости между случайными величинами. Они помогают измерить, насколько две случайные величины варьируются вместе. Ниже приведены основные ковариационные характеристики:
    - **Ковариация**. Пусть $(\Omega, F, P)$ - вероятностный эксперимент, а $\xi, \eta$ - случайные величины, определенные в нем. Смешанный центральный момент второго порядка $$\mathbb{E}(\xi-\mathbb{E}\xi)(\eta-\mathbb{E}\eta) = \mathop{\mathrm{cov}}\limits(\xi, \eta)$$ называется ковариацией случайных величин $\xi$ и $\eta$.
    - **Коэффициент корреляции**. Пусть $\xi,\eta$ - случайные величины, у которых существуют дисперсии и $\mathbb{D}\xi\mathbb{D}\eta > 0$. Тогда число $$\rho(\xi,\eta) = \frac{\mathop{\mathrm{cov}}\limits(\xi, \eta)}{\sqrt{\mathbb{D}\xi\cdot\mathbb{D}\eta}}$$
    - **Матрица ковариациий** - это матрица, составленная из попарных ковариаций элементов одного или двух случайных векторов. $$\vec{\xi}=\begin{pmatrix} 
    r_{11} & ... & r_{1n} \\
    ... & ... & ... \\
    r_{n1} & ... & r_{nn} 
    \end{pmatrix}, r_{ij} = \mathop{\mathrm{cov}}\limits(\xi_{i}, \xi_{j})$$ Элементы на главной диагонали представляют собой дисперсии каждой компоненты вектора.

### Задание 1
Вычислить вектор  математических ожиданий и ковариационные характеристики данного случайного вектора. Построить график плотности распределения $p_{\xi,\eta}$.

Преобразуем квадратичную форму в показателе экспоненты

$q(x,y) = 2x^2 - 4xy + 5y^2 - 8x + 2y + 11$

$2x^2 - 4xy + 5y^2 - 8x + 2y + 11 $\
$= 2(x^2 - 2xy - 4x) + 5y^2 + 2y + 11 $\
$= 2(x^2 - 2x(y + 2) + (y + 2)^2) - 2(y + 2)^2 + 5y^2 + 2y + 11 $\
$= 2(x - (y + 2))^2 - 3(y - 1)^2 = 2((x - 3) - (y - 1))^2 + 3(y - 1)^2 $\
$= 2(x - 3)^2 - 4(x - 3)(y - 1) + 5(y - 1)^2 $

$= 2(x - 3)^2 - 4(x - 3)(y - 1) + 5(y - 1)^2 $

По каноническому виду мы можем найти мат. ожидание $\mathbb{E}\vec{\xi}$ и матрицу ковариаций $ \Sigma $. \
$\mathbb{E} \begin{pmatrix} \xi \\ \eta \end{pmatrix} = \begin{pmatrix} 3 \\ 1 \end{pmatrix} \quad \Sigma^{-1} = \begin{pmatrix} 2 & -2 \\ -2 & 5\end{pmatrix} \quad det \Sigma^{-1} = 10 - 4 = 6 \implies \Sigma = \frac{1}{6} \begin{pmatrix} 5 & 2 \\ 2 & 2 \end{pmatrix}$ \
\
Найдём ковариационные характеристики данного случайного вектора
$ cov(\xi, \eta) = \frac{1}{3} \quad \mathbb{D}_{\xi} = \frac{5}{6}  \quad \mathbb{D}_{\eta} = \frac{1}{3}  \implies \rho(\xi, \eta)=\frac{cov({\xi},{ \eta})}{\sqrt{\mathbb{D}_{\xi} \mathbb{D}_{\eta}}} = \frac{\frac{1}{3}}{\sqrt{\frac{5}{6} * {\frac{1}{3}}}} = \frac{\sqrt{10}}{5}$ \
\
Найдём коэффициент плотности распределения $C$. \
$C = \frac{1}{(2 \pi)^{\frac{n}{2}} \sqrt{det \Sigma}} = \frac{1}{2 \pi \sqrt{\frac{1}{6}}} =\frac{3}{\pi \sqrt{6}}$ \
\
Тогда функция плотности выглядит следующим образом \
$\rho_{\xi, \eta}(x, y)= \frac{3}{\pi \sqrt{6}} \cdot \exp(-\frac{1}{2}(2x^2-4xy+5y^2-8x+2y+11)) $


```python
def p(x, y):
    return (1 / (np.pi *3 * np.sqrt(6))) * np.exp(-0.5 * (2 * x ** 2 - 4 * x * y + 5 * y ** 2 - 8 * x + 2 * y + 11))


x = np.linspace(-3, 6, 200)
y = np.linspace(-3, 5, 200)

X, Y = np.meshgrid(x, y)
Z = p(X, Y)

fig = plt.figure(figsize=(10,8))
ax = plt.axes(projection='3d')

ax.plot_surface(X, Y, Z)
ax.set_title('График функции плотности')
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('y', fontsize=12)
ax.set_zlabel('z', fontsize=12)
plt.show()
```
