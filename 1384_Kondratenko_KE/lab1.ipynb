## Основные теоретические положения
Случайный вектор $\overrightarrow{\xi} = (\xi_1, \dots, \xi_n)^{T}$ имеет многомерное нормальное распределение, если:

 $\overrightarrow{\xi} = A\overrightarrow{\eta} + \overrightarrow{b} $.

Плотность многомерного нормального распределения имеет вид: $p_{\overrightarrow{\xi}}\left(x_1, \dots, x_n\right) = \frac{1}{(2\pi)^{n/2}\sqrt{\det \Sigma}} \exp\left(-\frac{1}{2}\left(\overrightarrow{x}-\overrightarrow{\mu}\right)^\intercal \Sigma^{-1}\left(\overrightarrow{x}-\overrightarrow{\mu}\right)\right)$

где $\overrightarrow{\mu}  - мат \ ожадние \ случайного \ вектора $ ${\xi} $.

Матрица ковариации: $\Sigma = \{\sigma_{i,j}\}_{i,j}^n$, где $\sigma_{i,j} = \mathop{cov}(\xi_i,\xi_j)$.

Свойства многомерного нормального распределения:

1) Многомерное нормальное распределение устойчиво к линейным преобразованиям.

2) Линейная случайных величин из многомерного нормального распределения образует нормально распределенной.

3) Параметры многомерного нормального распределения полно описывают его форму.

4) Условные и безусловные распределения, полученные из многомерного нормального распределения, также являются нормально распределенными.

### 4. Вычислить характеристики совместного распределения случайного вектора $(-\xi - 2\eta, \xi - 2\eta)^T$ и записать его плотность. Построить график плотности полученного распределения.

$\begin{pmatrix} \xi_4 \\ \eta_4 \end{pmatrix} = \begin{pmatrix} -\xi-2\eta \\ \xi -2\eta \end{pmatrix}$

$ A = \begin{pmatrix} -1 &-2 \\ 1&-2 \end{pmatrix} \begin{pmatrix} \xi \\ \eta \end{pmatrix} $


$\det{A} = 4$


$E \begin{pmatrix} \xi_4 \\ \eta_4 \end{pmatrix} = \begin{pmatrix} -1 &-2 \\ 1&-2 \end{pmatrix} \begin{pmatrix} -1 \\ 1 \end{pmatrix} = \begin{pmatrix} -1 \\ -3 \end{pmatrix} $

$\Sigma_4 = A \Sigma A^T = \frac{2}{75} \begin{pmatrix} -1 &-2 \\ 1&-2 \end{pmatrix}\begin{pmatrix} 14 & -3 \\ -3 & 6 \end{pmatrix}\begin{pmatrix} -1 & 1 \\ -2&-2 \end{pmatrix} = \frac{2}{75} \begin{pmatrix} -8 & -9 \\ 20 & -15 \end{pmatrix}\begin{pmatrix} -1 & 1 \\ -2&-2 \end{pmatrix} = \frac{2}{75} \begin{pmatrix} 26 & 10 \\ 10 & 50 \end{pmatrix} = \frac{4}{75} \begin{pmatrix} 13 & 5 \\ 5 & 25 \end{pmatrix}$


$det{\Sigma_4} = \frac{64}{75} $

$D_\xi = \frac{13\cdot4}{75}$

$D_\eta = \frac{25\cdot4}{75} = \frac{4}{3}$

$cov(\xi, \eta) = \frac{5*4}{75} =\frac{4}{15} $

$\rho_{\xi, \eta} (x, y) = \frac{cov(\xi, \eta)}{\sqrt{D_\xi \cdot D_\eta}} = \frac{\frac{4}{15}}{\sqrt{\frac{13\cdot4\cdot4}{75\cdot3}}} = \frac{\frac{4}{15}}{\frac{4}{15}\sqrt{13}} = \frac{1}{\sqrt{13}}$

$\Sigma_4^{-1} = \frac{1}{\frac{64}{75}} \begin{pmatrix} \frac{4}{3} & -\frac{4}{15} \\ -\frac{4}{15} & \frac{52}{75}  \end{pmatrix} = \frac{75}{64}\begin{pmatrix} \frac{4}{3} & -\frac{4}{15} \\ -\frac{4}{15} & \frac{52}{75}  \end{pmatrix} $


$C = \frac{1}{2\pi\sqrt{det{\Sigma_4}}} = \frac{1}{\frac{2\pi\cdot8}{5\sqrt{3}}} = \frac{5\sqrt{3}}{16\pi}$


$p_{\xi,\eta}(x,y) =  \frac{5\sqrt{3}}{16\pi} \exp\left(-\frac{75}{128}\left[
  \begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix}-1 \\ -3  \end{pmatrix}\right]^{T} \begin{pmatrix} \frac{4}{3} & -\frac{4}{15} \\ -\frac{4}{15} & \frac{52}{75}  \end{pmatrix}  \left[ \begin{pmatrix}    x \\ y  \end{pmatrix} - \begin{pmatrix}  -1 \\ -3
  \end{pmatrix}\right]\right)$


```python
import math
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap

fig = plt.figure(figsize=(5,5))
ax_3d = fig.add_subplot(projection='3d')

x = np.linspace(-5, 3, 100)
y = np.linspace(-8, 3, 100)
x, y = np.meshgrid(x, y)
c = 5*math.sqrt(3)/(16*math.pi)
q = (75/64)*((4/3)*(x+1)**2 - (8/15)*(x+1)*(y+3) + (52/75)*(y+3)**2)
z = c*np.exp((-1/2)*q)
cmap = LinearSegmentedColormap.from_list ('oumuamua', ['white', '#a63232', '#ddf50a', 'blue'], 256)
image = ax_3d.plot_surface(x, y, z, cmap=cmap ,linewidth=0, antialiased=False)
plt.show(image)
```
