## Основные теоретические положения
Случайный вектор $\overrightarrow{\xi} = \left(\xi_1, \dots, \xi_n\right)^\intercal:\Omega\to \mathbb{R}^n$ имеет многомерное нормальное распределение, если существует вектор независимых стандартных нормальных случайных величин $\overrightarrow{\eta} = \left(\eta_1, \dots, \eta_m\right)^\intercal$, матрица A размерности $n \times m$ и вещественный вектор $\overrightarrow{b} = (b_1,\dots,b_n)^\intercal$, такие что: 

 $\overrightarrow{\xi} = A\overrightarrow{\eta} + \overrightarrow{b} $.


Многомерное случайное распределение - абсолютно непрерывное распределение, с плотностью:
$\mathcal{p}_{{\xi}}(x) = \frac{1}{(2\pi)^\frac{n}{2}\sqrt{\det \Sigma}} \exp\left(-\frac{1}{2}\left(\overrightarrow{x}-\overrightarrow{\mu}\right)^\intercal \Sigma^{-1}\left(\overrightarrow{x}-\overrightarrow{\mu}\right)\right)$

Его числовые характеристики:

Математическое ожидание: $\mathbb{E}\overrightarrow{\xi}=\mu$ 
Матрица ковариаций: $\left\{\sigma_{i,j}\right\}_{i,j=1}^n = \Sigma$, где $\sigma_{i,j} = \mathop{\mathrm{cov}}\left(\xi_i,\xi_j\right)$



Свойства многомерного нормального распределения:
1. Линейная комбинация случайных переменных, принадлежащих многомерному нормальному распределению, также является нормально распределенной.
2. Условные и безусловные распределения, полученные из многомерного нормального распределения, также являются нормально распределенными.
3. Параметры многомерного нормального распределения полностью описывают его форму.
4. Многомерное нормальное распределение устойчиво относительно линейных преобразований. Если $\overrightarrow{\xi} \sim \mathcal{N}\left(\overrightarrow{\mu}, \Sigma\right)$, $A$ - произвольная матрица размерности $m \times n$, $\overrightarrow{b}$ - произвольный вектор размерности $m \times 1$, то $A\overrightarrow{\xi}+\overrightarrow{b} \sim \mathcal{N}\left(A\overrightarrow{\mu}+\overrightarrow{b}, A \Sigma A^\intercal\right)$.

### Задание 1
Вычислить вектор математических ожиданий и ковариационнные характеристики данного случайного вектора. Построить график плотности распределения $p_{\xi, \eta}$.

Пусть $q(x,y) = 3x^2 - 4xy + 6y^2 + 14x + 21 $

Воспользуемся методом Лагранжа для приведения квадратичной формы к каноническому виду.

$q(x,y) = \left(3x^2 - 4xy + 14x\right) + 6y^2 + 21  = \left(3x^2 - 2x(2y - 7)\right) + 6y^2 + 21 = 3 \left( x^2 - 2x\cdot \frac{1}{3}(2y-7) \right) + 6y^2 + 21 = 3\left(x^2 - 2x \left(\frac{2y}{3} -\frac{7}{3}\right) + \left(\frac{2y}{3} -\frac{7}{3}\right)^2 - \left(\frac{2y}{3} -\frac{7}{3}\right)^2 \right) + 6y^2 + 21 = 3\left(x- \frac{2y}{3} +\frac{7}{3}\right)^2 - 3\left(\frac{2y}{3} -\frac{7}{3}\right)^2 + 6y^2 + 21 = 3\left(x- \frac{2y}{3} +\frac{7}{3}\right)^2 -\frac{4y^2}{3} +\frac{28y}{3} -\frac{49}{3}+ 6y^2 + 21 =  3\left(x- \frac{2y}{3} +\frac{7}{3}\right)^2 +\frac{14y^2}{3} +\frac{28y}{3} +\frac{14}{3} = 3\left(x+3\right)^2 -4(x+3)(y+1)+ \frac{4}{3}\left(y+1\right)^2 + \frac{14}{3}\left(y+1\right)^2 =  3\left(x+3\right)^2 -4(x+3)(y+1)+ 6\left(y+1\right)^2$

Вектор мат. ожиданий и обратная ковариационная матрица имеют вид:

$\mathbb{E} \begin{pmatrix} \xi \\ \eta \end{pmatrix} = \begin{pmatrix} -3 \\ -1 \end{pmatrix}$

$\Sigma^{-1} = \begin{pmatrix} 3 & -2 \\ -2 & 6 \end{pmatrix}$

Ее определитель будет равен $\det\left(\Sigma^{-1}\right) = 14$

Матрица ковариаций:

$\Sigma = \frac{1}{14} \begin{pmatrix} 6 & 2 \\ 2 & 3 \end{pmatrix}$

Дисперсии для каждой случайной величины равны:

$\mathbb{D}\xi = \frac{6}{14} = \frac{3}{7} \;\;\;\; \mathbb{D}\eta = \frac{3}{14}$

Найдём ковариацию и коэффициент корреляции:

$cov(\xi, \eta) = \frac{-2}{14}= \frac{-1}{7}$

$\rho(\xi, \eta) = \frac{\frac{-1}{7}}{\sqrt{\frac{1}{2}}\cdot \frac{3}{7}} = \frac{-\sqrt{2}}{3}$

Найдём константу C:

$C = \frac{1}{2\pi \sqrt{\frac{1}{14}}} =\frac{\sqrt{14}}{2\pi}$

Функция плотности распредления имеет вид:

$p_{\xi, \eta}(x,y) = \frac{\sqrt{14}}{2\pi} \cdot \exp \left(-\frac{1}{2} \begin{bmatrix}\begin{pmatrix} x \\ y \end{pmatrix}  - \begin{pmatrix} -3 \\ -1 \end{pmatrix} \end{bmatrix}^{\intercal} \cdot \begin{pmatrix} 3 & -2 \\ -2 & 6 \end{pmatrix} \cdot \begin{bmatrix}\begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} -3 \\ -1 \end{pmatrix} \end{bmatrix} \right)$

Построим график плотности распределения $p_{\xi, \eta}$


```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

def func(x, y):
    return (7 / (np.pi * np.sqrt(14))) * np.exp(-0.5 * (3 * x ** 2 - 4 * x * y + 6 * y ** 2 +14 * x + 21))


x = np.linspace(-5, -1, 200)
y = np.linspace(-4, 2, 200)

X, Y = np.meshgrid(x, y)
Z = func(X, Y)

fig = plt.figure(figsize=(10,5))
ax = plt.axes(projection='3d')

ax.plot_surface(X, Y, Z, cmap='magma')
ax.set_title('График плотности')
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('y', fontsize=12)
ax.set_zlabel('z', fontsize=12)
plt.show()
```

### Задание 3
Найти ортогональное преобразование, переводящее соответствующий центрированный случайный вектор в вектор с независимыми компонентами. Построить график плотности полученного распределения.

Найдем собственные векторы матрицы $\Sigma^{-1}$

$\Sigma^{-1} = \begin{pmatrix} 3 & -2 \\ -2 & 6 \end{pmatrix}$

Для этого найдём собственные числа

$ \begin{vmatrix} 3 - \lambda & -2 \\ -2 & 6 - \lambda \end{vmatrix} = 0$

$ \lambda_{1} = 2$

$\begin{pmatrix} 1 & -2 \\ -2 & 4 \end{pmatrix} \overrightarrow{s_{1}} = \begin{pmatrix} 0 \\ 0 \end{pmatrix}$

Получим нормированный собственный вектор:

$ \overrightarrow{s_{1}} = \frac{1}{\sqrt{5}}\begin{pmatrix} 2 \\ 1 \end{pmatrix}$

$ \lambda_{2} = 7$

$\begin{pmatrix} -4 & -2 \\ -2 & -1 \end{pmatrix} \overrightarrow{s_{2}} = \begin{pmatrix} 0 \\ 0 \end{pmatrix}$

Получим нормированный собственный вектор:

$ \overrightarrow{s_{2}} = \frac{1}{\sqrt{5}}\begin{pmatrix} -1 \\ 2 \end{pmatrix}$

Объединив векторы и транспонировав матрицу, получим матрицу ортогонального преобразования.

$ Q^{\intercal} = \frac{1}{\sqrt{5}} \begin{pmatrix} 2 & -1 \\ 1 & 2 \end{pmatrix}$

$ Q = \frac{1}{\sqrt{5}} \begin{pmatrix} 2 & 1 \\ -1 & 2 \end{pmatrix}$


$\begin{pmatrix} \xi_{ort} \\ \eta_{ort} \end{pmatrix} = Q \begin{pmatrix} \xi \\ \eta \end{pmatrix} = \frac{1}{\sqrt{5}} \begin{pmatrix} 2\xi + \eta \\ -\xi + 2\eta \end{pmatrix}$

$\mathbb{E} \begin{pmatrix} \xi_{ort} \\ \eta_{ort} \end{pmatrix} = \frac{1}{\sqrt{5}} \begin{pmatrix} 2\cdot(-3) + (-1) \\ -3 + 2\cdot(-1) \end{pmatrix} = \frac{1}{\sqrt{5}} \begin{pmatrix} -7 \\ 1 \end{pmatrix}$

$\Sigma_{ort} = \frac{1}{14} \cdot \frac{1}{5} \begin{pmatrix} 2 & 1 \\ -1 & 2 \end{pmatrix}  \begin{pmatrix} 6 & 2 \\ 2 & 3 \end{pmatrix} \begin{pmatrix} 2 & -1 \\ 1 & 2 \end{pmatrix} = \frac{1}{70} \begin{pmatrix} 14 & 7 \\ -2 & 4 \end{pmatrix} \begin{pmatrix} 2 & -1 \\ 1 & 2 \end{pmatrix} = \frac{1}{70} \begin{pmatrix} 35 & 0 \\ 0 & 10 \end{pmatrix} =  \begin{pmatrix}  \frac{1}{2} & 0 \\ 0 &  \frac{1}{7} \end{pmatrix} $

$\Sigma^{-1}_{ort} = \begin{pmatrix} 2 & 0 \\ 0 & 7 \end{pmatrix}$

$\mathbb{D}\xi_{ort} = \frac{1}{2} \;\;\;\; \mathbb{D}\eta_{ort} = \frac{1}{7} $

$cov(\xi_{ort}, \eta_{ort}) = 0 \;\; \rho(\xi_{ort}, \eta_{ort}) = 0$

Далее найдём C

$C = \frac{1}{2\pi \sqrt{\frac{1}{14}}} = \frac{\sqrt{14}}{2\pi}$

Функция плотности распредления имеет вид:


$p_{\xi_{ort}, \eta_{ort}}(x,y) = \frac{\sqrt{14}}{2\pi} \cdot \exp \left(-\frac{1}{2} \begin{bmatrix} \begin{pmatrix} x \\ y \end{pmatrix}  - \begin{pmatrix} \frac{-3}{\sqrt{5}} \\ \frac{-1}{\sqrt{5}} \end{pmatrix} \end{bmatrix}^{\intercal} \cdot \begin{pmatrix} 2 & 0 \\ 0 & 7 \end{pmatrix} \cdot \begin{bmatrix}\begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} \frac{-3}{\sqrt{5}} \\ \frac{-1}{\sqrt{5}} \end{pmatrix} \end{bmatrix} \right)$


```python
def func(x, y):
    return (np.sqrt(14) / (np.pi *2)) * np.exp(-0.5 * (2*x**2 +(28/np.sqrt(5))*x +21 + 7*y**2 -(14/np.sqrt(5))*y ))

x = np.linspace(-6, 0, 200)
y = np.linspace(-3, 3, 200)

X, Y = np.meshgrid(x, y)
Z = func(X, Y)

fig = plt.figure(figsize=(10,5))
ax = plt.axes(projection='3d')

ax.plot_surface(X, Y, Z, cmap='magma')
ax.set_title('График плотности')
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('y', fontsize=12)
ax.set_zlabel('z', fontsize=12)
plt.show()
```

$\begin{pmatrix} \xi_{4} \\ \eta_{4} \end{pmatrix} = \begin{pmatrix} -\xi- \eta \\ -\xi - 3\eta \end{pmatrix}$

$A = \begin{pmatrix} -1 & -1 \\ 1 & -3 \end{pmatrix}$

$\begin{pmatrix} \xi_{4} \\ \eta_{4} \end{pmatrix} = \begin{pmatrix} -1 & -1 \\ 1 & -3 \end{pmatrix} \begin{pmatrix} \xi \\ \eta \end{pmatrix}$

$\mathbb{E} \begin{pmatrix} \xi_{4} \\ \eta_{4} \end{pmatrix} =\begin{pmatrix} -1 & -1 \\ 1 & -3 \end{pmatrix} \begin{pmatrix} -3 \\ -1 \end{pmatrix} = \begin{pmatrix} 4 \\ 0 \end{pmatrix}$
